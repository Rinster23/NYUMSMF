```python
import pandas as pd
import numpy as np
from numpy import linalg as la
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

file_names = ["DGS1MO","DGS3MO","DGS6MO","DGS1","DGS2","DGS3","DGS5","DGS10","DGS20","DGS30"]
all_data = pd.DataFrame()
for i in range(len(file_names)):
    file_path = './datasets/' + file_names[i] + '.xlsx'
    data = pd.read_excel(file_path,header=10, usecols=[0,1])
    if i == 0:
        all_data = data
    else:
        all_data[file_names[i]] = data[file_names[i]]
all_data = all_data.fillna(method='pad')

### (C) ###
weekly_data = all_data.iloc[::7]
weekly_data.set_index('observation_date',drop=True,inplace=True)
train_data = np.array(weekly_data.loc["2001-08-01":"2005-12-31"])
test_data = np.array(weekly_data.loc["2006-01-01":])

change = train_data[1:,:] - train_data[:-1,:]
change_test = test_data[1:,:] - test_data[:-1,:]

pca = PCA(n_components=3)
pca.fit(change) # shape = (n_samples, n_features)
### (d) ###
plt.figure(figsize=(12, 5))
l1, = plt.plot(pca.components_[0])
l2, = plt.plot(pca.components_[1])
l3, = plt.plot(pca.components_[2])
plt.xlabel("Maturity")
plt.ylabel("Factor Loading")
plt.xticks(range(10),file_names)
plt.legend(handles=[l1,l2,l3],labels=['PC1','PC2','PC3'])
plt.show()
```

(d) \
PC1: Level: general level of interest rates, where all rates go up or down by the same amount. \
PC2: Steepness or slope: how rapidly the yield curve goes up (or down) from short rates to longer rates. \
PC3: Curvature or twist: butterfly shape that occurs between short, medium, and long-term rates.

```python
### (e) ###
V = pca.components_
result_test = V @ change_test.T
test_pca_var = np.var(result_test.T,axis=0)
test_var = np.var(change_test,axis=0)
print("Variance explained on training set: ", "%.4f" % sum(pca.explained_variance_ratio_))
print("Variance explained on test set: ",  "%.4f" % (sum(test_pca_var)/sum(test_var)))
```

(e) In the training set PCA explained more variance, possibly because we trained the PCA model on the training set, so it is normal that PCA had better performance on the training set.
