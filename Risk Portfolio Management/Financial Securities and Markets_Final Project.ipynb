```python
import yfinance as yf
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Define the tickers for the indices
tickers = [["^GSPC", "^STOXX50E", "^FTSE", "^SSMI"], 
           ["AAPL", "MSFT", "AMZN", "GOOGL", "GOOG", "META"],    
           ["NOVO-B.CO", "MC.PA", "ASML.AS", "OR.PA", "RMS.PA"],
           ["ULVR.L", "AZN.L", "HSBA.L", "SHEL.L", "BP.L"],
           ["NESN.SW", "ROG.SW", "UBSG.SW", "NOVN.SW", "CFR.SW"],
           ["^IRX", "^FVX", "^TNX","^TYX"],
           ["EURUSD=X", "GBPUSD=X", "JPYUSD=X", "CADUSD=X", "CHFUSD=X"],
           ["GC=F", "CL=F", "ZC=F"]]

# Set the start date and end date for the historical data
end_date = datetime.today().strftime('%Y-%m-%d')

# Fetch historical data for each group and use adjusted close prices
indices = yf.download(tickers[0], start='2013-01-01', end=end_date)["Adj Close"]
stocks_us = yf.download(tickers[1], start='2013-01-01', end=end_date)["Adj Close"]
stocks_eu = yf.download(tickers[2], start='2013-01-01', end=end_date)["Adj Close"]
stocks_br = yf.download(tickers[3], start='2013-01-01', end=end_date)["Adj Close"]
stocks_sw = yf.download(tickers[4], start='2013-01-01', end=end_date)["Adj Close"]
treasury_rate = yf.download(tickers[5], start='2013-01-01', end=end_date)["Adj Close"]
FX = yf.download(tickers[6], start='2013-01-01', end=end_date)["Adj Close"]
commodities = yf.download(tickers[7], start='2013-01-01', end=end_date)["Adj Close"]
```

```python
# Fill NaN values with the previous value
def calRet(df, absolute=False):
    df.fillna(method='bfill', inplace=True)
    if absolute:
        df = df - df.shift()
    else:
        df = df.pct_change()
    df.dropna(inplace=True)
    return df

indices = calRet(indices)
stocks_us = calRet(stocks_us)
stocks_eu = calRet(stocks_eu)
stocks_br = calRet(stocks_br)
stocks_sw = calRet(stocks_sw)
treasury_rate = calRet(treasury_rate, True)
FX = calRet(FX)
commodities = calRet(commodities)
```

```python
# (a)
def gen_corr(data, months=[1, 3, 6, 12]):
    end_date = pd.to_datetime(data.index.values[-1])
    for month in months:
        date = end_date - pd.DateOffset(months=month)
        print("Correlation {}M: \n".format(month), data[data.index >= date].corr())
print("Correlation between main indices")
gen_corr(indices)
```

```python
import matplotlib.pyplot as plt
# (b)
# calculate monthly correlations between each of these indices and 
# draw the time series 
def gen_corrs(data):
    dict_month_corr = {}
    monthly_data = data.resample('M')
    i = 1
    for _, group in monthly_data:
        dict_month_corr[i] = group.corr()
        i += 1
    return dict_month_corr

def plot_corr(dict_month_corr, plt_name):
    cols = dict_month_corr[1].columns.values
    lines = []
    names = []
    for i in range(len(cols)):
        for j in range(i+1, len(cols)):
            col1 = cols[i]
            col2 = cols[j]
            line = [data.loc[col1,col2] for data in dict_month_corr.values()]
            name = col1.replace('^','') + '&' + col2.replace('^','')
            lines.append(line)
            names.append(name)
    plt.figure(figsize=(16, 12))
    plot_lines = []
    for line in lines:
        line, = plt.plot(line)
        plot_lines.append(line)
    plt.xlabel("Month")
    plt.ylabel("Correlation")
    plt.title("Time series for {} correlations".format(plt_name))
    plt.legend(handles=plot_lines,labels=names) 
    plt.show()

    
indices_corr = gen_corrs(indices)
plot_corr(indices_corr, "main indices")
```

#### (c) The main indices are positively correlated under most cases, and they tend to have synchronous large oscillation, possibly during periods when global markets were volatile.

```python
# (b)
def plot_ret(data, months=[1, 3, 6, 12, 24, 60, 120]):
    end_date = pd.to_datetime(data.index.values[-1])
    # Create a 7x4 grid of subplots
    fig, axes = plt.subplots(nrows=len(months), ncols=len(data.columns), figsize=(12, 18))  # Adjust figsize as needed
    # Iterate through each subplot
    for i in range(len(months)):
        date = end_date - pd.DateOffset(months=months[i])
        temp_data = data[data.index >= date]
        for j in range(len(data.columns)):
            ax = axes[i, j]
            plot_data = temp_data.iloc[:,j]
            ax.hist(
            plot_data,
            color="lightblue",
            edgecolor="grey",
            bins=50,
            )
            ax.set_title("{} in {} mon".format(data.columns[j], months[i]))
            ax.set_xlabel('Return')
            ax.set_ylabel('Frequency')

    fig.suptitle('Histograms of return', fontsize=16)
    fig.tight_layout()
    plt.show()
            
plot_ret(indices)    
```

```python
# (c)
# use jarque_bera to test normality for return in past 11 years
from scipy import stats
def JB_test(data, obj):
    for i in range(len(data.columns)):
        test_data = data.iloc[:,i]
        res = stats.jarque_bera(test_data)
        if res.pvalue < 0.1:    # 90% confidence level 
            print("Reject H_0: {} of {} follows normal distribution".format(obj, data.columns[i]))
        else:
            print("Cannot reject H_0: {} of {} follows normal distribution".format(obj, data.columns[i]))

JB_test(indices, "Return")
```

#### From JB test, we reject the hypothesis that return of each equity index follows normal doistribution based on historical data.

#### From the histogram, we can see that the return distribution is highly leptokurtic, so we can fit t distribution.

#### It is also in correspondence with empirical finance: return distributions are usually highly leptokurtic.

```python
# (d)
import warnings
warnings.filterwarnings("ignore")

def plot_stat(data, months=[1, 3, 6, 12]):
    days = 21
    temp_data = data.copy()
    cols = data.columns
    for col in cols:
        for month in months:
            temp_data['var_{}_{}'.format(col, month)] = temp_data[col].rolling(month*days).var()
            temp_data['skew_{}_{}'.format(col, month)] = temp_data[col].rolling(month*days).skew()
            temp_data['kurt_{}_{}'.format(col, month)] = temp_data[col].rolling(month*days).kurt()
    names = data.columns
    var_data = temp_data[[i for i in temp_data.columns if 'var' in i]]
    dict_var = {}
    for name in names:
        temp_var_data = var_data[[i for i in var_data.columns if name in i]]
        temp_var_data.dropna(inplace=True)
        dict_var[name] = temp_var_data.mean(axis=1)
    skew_data = temp_data[[i for i in temp_data.columns if 'skew' in i]]
    dict_skew = {}
    for name in names:
        temp_skew_data = skew_data[[i for i in skew_data.columns if name in i]]
        temp_skew_data.dropna(inplace=True)
        dict_skew[name] = temp_skew_data.mean(axis=1)
    kurt_data = temp_data[[i for i in temp_data.columns if 'kurt' in i]]
    dict_kurt = {}
    for name in names:
        temp_kurt_data = kurt_data[[i for i in kurt_data.columns if name in i]]
        temp_kurt_data.dropna(inplace=True)
        dict_kurt[name] = temp_kurt_data.mean(axis=1)
    # plot
    for my_dict, title in [(dict_var,"variance"), (dict_skew,"skewness"), (dict_kurt,"kurtosis")]:
        lines = []
        plt.figure(figsize=(12, 5))
        for col in data.columns:
            line, = plt.plot(my_dict[col].values)
            lines.append(line)
        plt.legend(handles=lines,labels=list(data.columns))
        plt.title("Rolling statistics for the average {}".format(title))
        plt.show()
plot_stat(indices)
```

```python
# (a)
import pandas as pd
import numpy as np
from datetime import datetime
import yfinance as yf
from numpy.lib.stride_tricks import sliding_window_view
import statsmodels.api as sm
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

df = pd.read_html("https://disfold.com/stock-index/sp-500/companies/")[0]
df.drop(9,inplace=True)
df.reset_index(drop=True)
tickers = [ticker.split(" ")[0] for ticker in list(df['Stock'])]
tickers[8] = "BRK-B"
tickers.remove("GOOG") # remove multicolinearity
tickers = ["^SPX"] + tickers
end_date = datetime.today().strftime('%Y-%m-%d')
snp_50 = yf.download(tickers, start='2019-01-01', end=end_date)["Adj Close"]
snp_50.fillna(method='bfill', inplace=True)

def rolling_regression(data,index_name,top5, months=3):
    days = 21
    temp_data = data.copy()
    cols = list(temp_data.columns)
    sep_data = sliding_window_view(temp_data, (days*months, len(cols))).squeeze()
    r2_list = []
    F_val = []
    betas = []
    for sub in sep_data:
        X = sm.add_constant(sub[:,:-1], has_constant='add')
        model = sm.OLS(sub[:,-1], X)
        result = model.fit()
        r2_list.append(result.rsquared)
        F_val.append(result.f_pvalue)
        betas.append(list(result.params))
    plt.figure(figsize=(12, 5))
    plt.plot(r2_list)
    plt.title(("Time series of R^2, regression on {} mon, "+index_name).format(months))
    plt.ylabel("R square")
    plt.show()
    plt.figure(figsize=(12, 5))
    plt.plot(F_val)
    plt.title(("Time series of p-value of F stat, regression on {} mon, "+index_name).format(months))
    plt.ylabel("p-value")
    plt.show()
    betas = np.array(betas)
    plt.figure(figsize=(12, 5))
    lines = []
    for i in range(5):
        top = top5[i]
        ind = cols.index(top)
        line, = plt.plot(betas[:,ind+1])
        lines.append(line)
    plt.legend(handles=lines,labels=top5)
    plt.ylabel("Betas")
    plt.title("Time series of beta for the top 5 underlyings of "+ index_name)
    plt.show()
    
rolling_regression(snp_50, "S&P500", top5=tickers[1:6])
```

#### $R^2$ is very close to 1, which means that our linear model fits the data well. The p-value of F-stat is very close to 0, which means that our model as a whole is significant. There aren't any pattern in the betas of the top 5 underlyings, since a stock's market cap is usually not proportional to its price.

```python
# (a)
from numpy.lib.stride_tricks import sliding_window_view

tickers = [["^GSPC", "^STOXX50E", "^FTSE", "^SSMI"], 
           ["AAPL", "MSFT", "AMZN", "GOOGL", "GOOG", "META"],    
           ["NOVO-B.CO", "MC.PA", "ASML.AS", "OR.PA", "RMS.PA"],
           ["ULVR.L", "AZN.L", "HSBA.L", "SHEL.L", "BP.L"],
           ["NESN.SW", "ROG.SW", "UBSG.SW", "NOVN.SW", "CFR.SW"],
           ["^IRX", "^FVX", "^TNX","^TYX"],
           ["EURUSD=X", "GBPUSD=X", "JPYUSD=X", "CADUSD=X", "CHFUSD=X"],
           ["GC=F", "CL=F", "ZC=F"]]
indices = yf.download(tickers[0], start='2013-01-01', end=end_date)["Adj Close"]
treasury_rate = yf.download(tickers[5], start='2013-01-01', end=end_date)["Adj Close"]
FX = yf.download(tickers[6], start='2013-01-01', end=end_date)["Adj Close"]
treasury_rate.fillna(method='bfill', inplace=True)
FX.fillna(method='bfill', inplace=True)


def calc_vol(data, log=False):
    if log:
        return np.sqrt(np.mean(np.square(np.log(data.squeeze())[:,1:,:] - np.log(data.squeeze())[:,:-1,:]), axis=1)*252)
    else:
        return np.sqrt(np.mean(np.square(data.squeeze()[:,1:,:] - data.squeeze()[:,:-1,:]), axis=1)*252)

def calc_var(data, log=False):
    return np.var(data.squeeze(), axis=1)
    
def get_stat(data, stat, months=1, log=False, plot=False, need=False):
    days = 21
    temp_data = data.copy()
    cols = temp_data.columns
#     res = []
#     for month in months[::-1]:
#         sep_data = sliding_window_view(temp_data, (days*month, len(cols)))
#         res.append(calc_func(sep_data, log=log))
#     length = min([i.shape[0] for i in res])
#     res = np.array([i[-length:,:] for i in res])
#     res = np.mean(res, axis=0)
    sep_data = sliding_window_view(temp_data, (days*months, len(cols)))
    if stat == "volatility":
        res = calc_vol(sep_data, log=log)
        if log:
            rets = res[1:,:] - res[:-1,:]
        else:
            rets = pd.DataFrame(res).pct_change()
            rets.dropna(inplace=True)
            rets = np.array(rets)
    else:
        rets = calc_var(sep_data, log=log)
        rets = pd.DataFrame(rets).pct_change()
        rets.dropna(inplace=True)
        rets = np.array(rets)
    if plot: 
        for i in range(rets.shape[1]):
            name = data.columns[i]
            plt.figure(figsize=(10, 5))
            plt.hist(
                rets[:,i],
                color="lightblue",
                edgecolor="grey",
                bins=100,
                )
            plt.title("Histogram for {} of {} mon, {}".format(('log 'if log else '')+stat,months,name))
            plt.show()
    if need:
        return rets

get_stat(indices, "volatility", plot=True)
JB_test(pd.DataFrame(get_stat(indices,"volatility", need=True),
                     columns = indices.columns), "realized volatility")
```

```python
# (a)
end_date = datetime.today().strftime('%Y-%m-%d')
spx_vix = yf.download(["^SPX","^VIX"], start='2013-01-01', end=end_date)["Adj Close"]
spx_vix_corr = gen_corrs(spx_vix)
plot_corr(spx_vix_corr, "SPX & VIX")
```

#### It is expected as under normal market conditions, there is often an inverse relationship between SPX and VIX. VIX is often referred to as the "fear gauge" because it tends to rise when there is fear or uncertainty in the market. As such, it can act as a leading indicator of potential market downturns. Therefore, when the stock market (SPX) rises, volatility (VIX) tends to fall, and vice versa. As we can see in the above diagram, the correlation are negative under most cases.

```python
# (b)
# calculate log-vol
SPX = spx_vix['^SPX']
def calc_vol_series(data, log=False):
    if log:
        return np.sqrt(np.mean(np.square(np.log(data)[1:,:] - np.log(data)[:-1,:]), axis=1)*252)
    else:
        return np.sqrt(np.mean(np.square(data[1:,:] - data[:-1,:]), axis=1)*252)

def get_stat_series(data, func, months=1, log=False):
    days = 21
    temp_data = data.copy()
    sep_data = sliding_window_view(temp_data, days*months)
    res = func(sep_data, log=log)
    return res

log_vol = get_stat_series(SPX, calc_vol_series, months=1, log=True)
plt.figure(figsize=(10, 5))
l1, = plt.plot(list(spx_vix['^VIX'].iloc[20:]))
l2, = plt.plot(log_vol*100)
plt.xlabel("Month")
plt.ylabel("Volatility")
plt.title("Time series for Volatility")
plt.legend(handles=[l1,l2],labels=["VIX", "Log-volatility"]) 
plt.show()
```

```python
# calculate Parkinson's 
SPX = yf.download(["^SPX"], start='2013-01-01', end=end_date)
high_low = SPX[['High','Low']]

def Parkinson(data):
    return np.sqrt(np.mean(np.square(np.log(data[:,:,1]/data[:,:,0])), axis=1)*252/4/np.log(2))

def get_stat_q5(data, func, months=1):
    days = 21
    temp_data = data.copy()
    sep_data = sliding_window_view(temp_data, (days*months,temp_data.shape[1]))
    sep_data = sep_data.squeeze()
    res = func(sep_data)
    return res


Parkinson_vol = get_stat_q5(high_low, Parkinson)
plt.figure(figsize=(10, 5))
l1, = plt.plot(list(spx_vix['^VIX'].iloc[20:]))
l2, = plt.plot(Parkinson_vol*100)
plt.xlabel("Month")
plt.ylabel("Volatility")
plt.title("Time series for Volatility")
plt.legend(handles=[l1,l2],labels=["VIX", "Parkinson-volatility"]) 
plt.show()
```

```python
hlo = SPX[['High','Low','Open', "Adj Close"]]
def Garman_Klass(data):
    return np.sqrt(np.mean(1/2*np.square(np.log(data[:,:,0]/data[:,:,1]))-
                           (2*np.log(2)-1)*np.square(np.log(data[:,:,3]/data[:,:,2])), axis=1)*252)

Garman_Klass_vol = get_stat_q5(hlo, Garman_Klass)
plt.figure(figsize=(10, 5))
l1, = plt.plot(list(spx_vix['^VIX'].iloc[20:]))
l2, = plt.plot(Garman_Klass_vol*100)
plt.xlabel("Month")
plt.ylabel("Volatility")
plt.title("Time series for Volatility")
plt.legend(handles=[l1,l2],labels=["VIX", "Garman_Klass-volatility"]) 
plt.show()
```

```python
# (c)
names = ["log-vol","Parkinson_vol", "Garman_Klass_vol"]
data = [log_vol, Parkinson_vol[1:], Garman_Klass_vol[1:]]
for i in range(len(names)):
    plt.figure(figsize=(6, 6))
    plt.scatter(list(spx_vix['^VIX'].iloc[:-21]), data[i]*100, s=0.5)
    plt.xlabel("VIX")
    plt.plot(np.linspace(5,80,1000),np.linspace(5,80,1000),c="r")
    plt.ylabel(names[i])
    plt.title("Scatter plot")
    plt.show()
```

```python
import yfinance as yf
from scipy.stats import norm
from scipy.optimize import fsolve
import numpy as np
from datetime import datetime, timedelta
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

def black_scholes(S, K, T, r, sigma):
    """
    Calculate the Black-Scholes option price for call options.

    Parameters:
    S (float): Current stock price
    K (float): Option strike price
    T (float): Time to expiration (in years)
    r (float): Risk-free interest rate
    sigma (float): Volatility of the underlying stock
    Returns:
    float: Black-Scholes option price
    """
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    option_price = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    return option_price

def calc_IV(option_price, spot_price, strike, time_to_maturity, r):
    def obj_func(sigma):
        return black_scholes(spot_price, strike, time_to_maturity, r, sigma) - option_price
    implied_vol = fsolve(obj_func, 0.5) 
    return implied_vol

def plot_vol_smile(ticker_symbol,expiration_dates):
    ticker = yf.Ticker(ticker_symbol)
    options_data = ticker.options
    tenors = ["1M", "3M", "6M", "9M", "1Y", "2Y", "3Y"]
    today = datetime.today().strftime('%Y-%m-%d')
    r = yf.download('^IRX', start=today, end=today)['Close'].values[0]*0.01 # today's 1Y treasury rate as risk free rate
    S0 = ticker.history(start=today, end=today)["Close"].values[0] # today's index value 
    today = datetime.strptime(today, "%Y-%m-%d")
    for i in range(len(expiration_dates)):
        IVs = []
        exp_date = expiration_dates[i]
        options_data = ticker.option_chain(exp_date)
        df = options_data.calls
        available_K = sorted(df['strike'])
        exp_date = datetime.strptime(exp_date, "%Y-%m-%d")
        ttr = (exp_date - today).days/365
        for K in available_K:
            opt_price = df[df['strike'] == K]["lastPrice"].values[0]
            # ans = df[df['strike'] == K]["impliedVolatility"].values[0]
            IVs.append(calc_IV(opt_price,S0,K,ttr,r))
        plt.figure(figsize=(10, 5))
        plt.plot(available_K, IVs)
        plt.xlabel("Strikes")
        plt.ylabel("Implied volatility")
        plt.title("Volatility smile for tenor {}, options on {}".format(tenors[i], ticker_symbol))
        plt.show()
        

expr_spx = ['2024-01-08', '2024-03-15', '2024-06-21', '2024-09-20','2024-12-20','2025-12-19', '2026-12-18']
plot_vol_smile("^SPX", expr_spx)
```

# Q8

The risk neutral density can be approximated with a butterfly spread,
$$
f(K) = e^{rT}\frac{c_1+c_3-2c_2}{\delta^2},
$$
where $c_i$ is the price of option with strike $K_i$ and $K_3-K_2 = K_2-K_1=\delta$.

```python
import yfinance as yf
from scipy.stats import norm
import numpy as np
from datetime import datetime, timedelta
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.nonparametric.kernel_regression import KernelReg
import pdb 

def rn_density(ticker_symbol, expiration_dates, plot=False):
    tenors = ["1M", "3M", "9M"]
    today = datetime.today().strftime('%Y-%m-%d')
    r = yf.download('^IRX', start=today, end=today)['Close'].values[0]*0.01 # today's 1Y treasury rate as risk free rate
    today = datetime.strptime(today, "%Y-%m-%d")
    ticker = yf.Ticker(ticker_symbol)
    for i in range(len(expiration_dates)):
        exp_date = expiration_dates[i]
        options_data = ticker.option_chain(exp_date)
        df = options_data.calls
        available_K = np.array((df['strike']))
        df['bidAskAvg'] = df[['bid','ask']].mean(axis=1)
        opt_prices = list(df["bidAskAvg"])
        diff = available_K[1:] - available_K[:-1]
        exp_date = datetime.strptime(exp_date, "%Y-%m-%d")
        ttr = (exp_date - today).days / 365
        density = []
        K_list = []
        for j in range(len(diff) - 1):
            if diff[j] < 50:
                if diff[j + 1] == diff[j]:
                    delta = diff[j]
                    den = abs(opt_prices[j] + opt_prices[j + 2]-2*opt_prices[j + 1])/delta/delta*np.exp(ttr*r)
                    density.append(den)
                    K_list.append(available_K[j + 1])
        model = KernelReg(density,K_list,var_type="c")
        res , _ = model.fit()
#         if stats:
#             n = len(K_list)
#             mean = sum(res[i]*K_list[i] for i in range(n))
#             var = sum([res[i]*K_list[i]**2 for i in range(n)])-mean**2
#             std = np.sqrt(var)
#             skew = n/(n-1)/(n-2)*sum(((i-mean)/std)**3 for i in range(n))
#             kur = n*(n+1)/(n-1)/(n-2)/(n-3)*sum(((i-mean)/std)**4 -3*(n-1)**2/(n-2)/(n-3) for i in range(n))
#             print("Mean for tenor {} is {}".format(tenors[i], mean))
#             print("Variance for tenor {} is {}".format(tenors[i], var))
#             print("Skewness for tenor {} is {}".format(tenors[i], skew))
#             print("Kurtosis for tenor {} is {}".format(tenors[i], kur))
        if plot:
            plt.figure(figsize=(10, 5))
            plt.plot(K_list, res)
            plt.xlabel("Index")
            plt.ylabel("Implied density")
            plt.title("Implied density for tenor {}, options on {}".format(tenors[i], ticker_symbol))
            plt.show()

rn_density("^SPX", ['2024-01-08', '2024-03-15', '2024-09-20'], plot=True)
```
