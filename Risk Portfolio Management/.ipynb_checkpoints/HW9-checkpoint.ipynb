# Q1

```python
import pandas as pd
import qrpm_funcs as qf
import numpy as np
from tabulate import tabulate
import matplotlib.pyplot as plt
import scipy.stats as spst

plt.rcParams['figure.dpi']= 300

lastday=qf.LastYearEnd()
#Swiss franc, pound sterling, Japanese Yen
seriesnames=['DEXSZUS','DEXUSUK','DEXJPUS']
cdates,ratematrix=qf.GetFREDMatrix(seriesnames,enddate=lastday)
multipliers=[-1,1,-1]

lgdates,difflgs=qf.levels_to_log_returns(cdates,ratematrix,multipliers)
```

```python
ind = lgdates.index('2020-12-31')
L = np.array(difflgs[:ind+1]) #Learning set
V = np.array(difflgs[ind+1:]) #Hold-out set

#Show histogram of equal-weighted
#CHF-GBP-JPY log-changes
w=np.array([1/3]*3).T
#apply transform to get returns at portfolio level
portfolio=np.log(1+np.dot(np.exp(L)-1,w))

#portfolio now contains the equal-weighted portfolio's
#log-returns. Create buckets - hist function doesn't
#seem to have "x or less" or "x or more" buckets
bucketnames=[]
bucketcounts=[]
#Bucket the end percentiles
low1=np.percentile(portfolio,1)
high1=np.percentile(portfolio,99)
bucketnames.append('<=%7.4f' % low1)
bucketcounts.append(sum(1 for x in portfolio if low1 >= x))
#Count 20 even buckets in between low and high
nbucket=20
bucketwidth=(high1-low1)/nbucket
for i in range(20):
    attach=low1+i*bucketwidth
    detach=attach+bucketwidth
    mid=(attach+detach)/2
    bucketnames.append('%7.4f' % mid)
    bucketcounts.append(sum(1 for x in portfolio if (x>=attach and x<=detach)))

#last bucket
bucketnames.append('>=%7.4f' % high1)
bucketcounts.append(sum(1 for x in portfolio if x >= high1))

width=.5
plt.bar(range(nbucket+2),bucketcounts,width)
plt.xticks(range(nbucket+2),bucketnames,rotation='vertical')
plt.grid()
plt.xlabel('log-change')
plt.ylabel('Day count')
plt.title('Histogram of equal-weighted CHF+GBP+JPY\n daily log-changes, '+ \
          lgdates[0][:4]+'-'+lgdates[ind][:4])
plt.show();

statnames,metrics,table=qf.StatsTable(np.exp(portfolio)-1)
print("std under equal-weight is ", table[5][1])
print("excess kurtosis under equal-weight is ", table[7][1])
```

```python
import math 

T = len(L)
h = 1/252
p_list = np.array([math.pow(2,-h*(i-1))*(1-math.pow(2,-h))/(1-math.pow(2,-h*T)) for i in range(1,T+1)])[::-1]
nums = p_list*T

#Redo histogram with EWMA
bucketnames=[]
bucketcounts=[]
#Bucket the end percentiles
low1 = np.percentile(portfolio,1)
high1 = np.percentile(portfolio,99)
bucketnames.append('<=%7.4f' % low1)
bucketcounts.append(sum([nums[i] for i in range(T) if low1 >= portfolio[i]]))
#Count 20 even buckets in between low and high
nbucket=20
bucketwidth=(high1-low1)/nbucket
for i in range(20):
    attach=low1+i*bucketwidth
    detach=attach+bucketwidth
    mid=(attach+detach)/2
    bucketnames.append('%7.4f' % mid)
    bucketcounts.append(sum([nums[i] for i in range(T) if detach>=portfolio[i]>=attach]))

#last bucket
bucketnames.append('>=%7.4f' % high1)
bucketcounts.append(sum([nums[i] for i in range(T) if portfolio[i] >= high1]))

width=.5
plt.bar(range(nbucket+2),bucketcounts,width)
plt.xticks(range(nbucket+2),bucketnames,rotation='vertical')
plt.grid()
plt.xlabel('log-change')
plt.ylabel('Day count')
plt.title('Histogram of EWMA CHF+GBP+JPY\n daily log-changes, '+ \
          lgdates[0][:4]+'-'+lgdates[ind][:4])
plt.show();
#EWMA std and excess kurtosis
mu = sum([portfolio[i] * p_list[i] for i in range(T)])
std = np.sqrt(sum([portfolio[i]**2 * p_list[i] for i in range(T)]) - mu**2)
print("std under EWMA is ", std)
kur = sum([((portfolio[i] - mu)/std)**4 * p_list[i] for i in range(T)]) - 3
print("excess kurtosis under EWMA is ", kur)
```

```python
#sample standard deviation and sample kurtosis of the equal-weighted portfolio in V
from scipy import stats
ret = np.dot(np.exp(V)-1,w)
print("std on hold-out set is ", np.std(ret))
print("excess kurtosis on hold-out set is ", stats.kurtosis(ret))
```

From the above results, we can see that EWMA does a better job in predicting the standard deviation on the hold out set, but both methods cannot give a good estimate of the excess kurtosis on the hold-out set.

Then take derivative of both sides with respect to $\lambda$, we have
$$

$$

```python
import numpy as np
import matplotlib.pyplot as plt

def func(x):
    return np.sin(x)**2*np.sin(2*x)**2*np.exp(-x*x/2)/np.sqrt(2*np.pi)

def M_H(f, x_0, itera, a, burn_in):
    res = []
    for i in range(itera):
        s = np.random.uniform(x_0-a, x_0+a)
        alpha = f(s)/f(x_0)
        if alpha >= np.random.uniform(0,1):
            if i >= itera * burn_in:
                res.append(s)
                x_0 = s
    plt.figure(figsize=(10, 5))
    plt.hist(
    res,
    color="lightblue",
    edgecolor="grey",
    bins=50,
    )
    plt.title("Histogram for samples from a = {}".format(a))
    plt.show()
```

```python
a_list = [i*0.1 for i in range(1,11)]
for a in a_list:
    M_H(func,3,10000,a,0.1)
```

```python
x = np.linspace(-3,3,10000)
y = func(x)
plt.figure(figsize=(10, 5))
plt.plot(x,y)
plt.title("True Distribution")
plt.show()
```

We can see from the above diagrams that $a = 1$ gives the most accurate sample. If $a$ is too small, then it would stuck in locality and hence fail to see the whole picture of the distribution.

```python
#  Levenes test

import qrpm_funcs as qf
import numpy as np


def signif_print(thresh,str_name,str_type,p_value):
#Print out text explaining significance
    one_minus_thresh=(1-thresh)*100

    if p_value < thresh:
        str_p=str_name+"Reject null hypothesis of equal "+str_type
        str_p+=" at %2.f" % one_minus_thresh
    else:
        str_p=str_name+"Cannot reject null hypothesis of equal "+str_type
        str_p+=" at %2.f" % one_minus_thresh
    str_p+="% significance"
    print(str_p)
    
    
lastday=qf.LastYearEnd()
#Swiss franc, pound sterling, Japanese Yen
seriesnames=['DEXSZUS','DEXUSUK','DEXJPUS']
cdates,ratematrix=qf.GetFREDMatrix(seriesnames,enddate=lastday)
multipliers=[-1,1,-1]

lgdates,difflgs=qf.levels_to_log_returns(cdates,ratematrix,multipliers)

#Mean vector and covariance matrix are inputs to efficient frontier calculations
d=np.array(difflgs)
m=np.mean(d,axis=0)
c=np.cov(d.T)

#**************************************************************
chol=np.linalg.cholesky(c)
count=len(d)

#Generate random draws; use fixed seed to be replicable
from numpy.random import default_rng
rng = default_rng(12345678)

s_trial=rng.normal(0,1,size=[count,3])
logr_trial=np.matmul(chol,s_trial.T).T+m
```

```python
from scipy import stats
difflgs = np.array(difflgs)
for i in range(3):
    res = stats.levene(difflgs[:,i],logr_trial[:,i],center='mean')
    p_value = res.pvalue
    signif_print(0.01,seriesnames[i]+": ","variances",p_value)
```

```python
# Box M test

def BoxM(T1,T2,s1,s2):
    #Tests for equality of two covariance matrices, s1 and s2
    #T1 and T2 are numbers of observations for s1 and s2
    #Returns M statistic and p-value

    #Make sure dimension is common
    if len(s1)!=len(s2):
        print("Error: different dimensions in Box M Test:",len(s1),len(s2))
        return(0,0)
    
    #Matrices are pxp
    p=len(s1)

    #Form the combined matrix
    scomb=(T1*s1+T2*s2)/(T1+T2)

    #Box M statistic
    Mstat=(T1+T2-2)*np.log(np.linalg.det(scomb))-(T1-1)* \
        np.log(np.linalg.det(s1))-(T2-1)*np.log(np.linalg.det(s2))

    #Multipliers from equation (49) in Box 1949.
    A1=(2*p**2+3*p-1)/(6*(p+1))
    A1*=(1/(T1-1)+1/(T2-1)-1/(T1+T2-2))

    A2=(p-1)*(p+2)/6
    A2*=(1/(T1-1)**2+1/(T2-1)**2-1/(T1+T2-2)**2)

    discrim=A2-A1**2

    #Degrees of freedom
    df1=p*(p+1)/2

    if discrim <= 0:
        #Use chi-square (Box 1949 top p. 329)
        test_value=Mstat*(1-A1)
        p_value=1-spst.chi2.cdf(test_value,df1)
    else:
        #Use F Test (Box 1949 equation (68))
        df2=(df1+2)/discrim
        b=df1/(1-A1-(df1/df2))
        test_value=Mstat/b
        p_value=1-stats.f.cdf(test_value,df1,df2)
    
    return(test_value,p_value)

#Apply to sample variances
stat, p_value = BoxM(count,count,c,np.cov(logr_trial.T))
print("Box M-stat and p-value:",stat,p_value)
signif_print(0.01,"","covariance matrices",p_value)
```
