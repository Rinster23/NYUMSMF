```python
import numpy as np
from scipy import stats
from scipy.stats import chi2
for _ in range(20):
    z = np.random.normal(loc=0.0,scale=1.0,size=(1000,1000))
    maximums = z.max(axis=1)
    res = stats.jarque_bera(maximums)
    #sta = res.statistic
    #print(1-chi2.cdf(sta,2) == res.pvalue)
    if res.pvalue < 0.025: # two sided 95% confidence level
        print("Reject H_0: X follows normal distribution")
    else:
        print("Cannot reject H_0")
```

Therefore, mx(i) also follows normal distribution. 

# Q3

```python
#chap4.py
import pandas as pd
import qrpm_funcs as qf
#Get 3 currencies until the end of
#previous year. Form sample covariance matrix
#and do simple efficient frontier calculations

lastday=qf.LastYearEnd()
#Swiss franc, pound sterling, Japanese Yen
seriesnames=['DEXSZUS','DEXUSUK','DEXJPUS']
cdates,ratematrix=qf.GetFREDMatrix(seriesnames,enddate=lastday)
multipliers=[-1,1,-1]

lgdates,difflgs=qf.levels_to_log_returns(cdates,ratematrix,multipliers)
```

```python
from scipy.stats import kurtosis
import math
portf_logRet = []
for i in difflgs:
    portf_logRet.append(np.log(np.mean(np.exp(i))))
portf_logRet = np.array(portf_logRet)*100 # in %
sigma_m = np.std(portf_logRet)
k_m = kurtosis(portf_logRet)
r_list = []
sigma2_list = []
for i in range(1,8):
    w_1 = 0.05 * i
    a = (k_m+3)/3*w_1**2-w_1
    b = 2/3*(k_m+3)*w_1*(1-w_1)
    c = (k_m+3)/3*(1-w_1)**2+w_1-1
    r = (-b-math.sqrt(b**2-4*a*c))/2/a
    sigma2 = sigma_m/math.sqrt(w_1*r+1-w_1)
    r_list.append(r)
    sigma2_list.append(sigma2)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
w1_list = [0.05 * i for i in range(1,8)]
plt.scatter(w1_list, r_list)
plt.scatter(w1_list, sigma2_list)
plt.xlabel("w_1")
plt.ylabel("Value")
plt.legend(labels=['r','sigma2'],loc='best')
plt.show()
```

# Q4

```python
from scipy.stats import norm
sorted_ret = np.sort(portf_logRet)
scores = []
mu = np.mean(portf_logRet)
n = len(sorted_ret)
for i in range(7):
    w_1 = w1_list[i]
    r = r_list[i]
    sigma2 = sigma2_list[i]
    sigma1 = math.sqrt(r)*sigma2
    obj = 0
    for j in range(n):
        cdf = w_1*norm.cdf(sorted_ret[j],loc=mu,scale=sigma1)+\
              (1-w_1)*norm.cdf(sorted_ret[j],loc=mu,scale=sigma2)
        obj += ((j+1)/(n+1) - cdf)**2
    scores.append(obj)

ind = scores.index(min(scores))
print("w_1 that minimizes objective function is: ", round(w1_list[ind],4))
```

```python
x=[i/(n+1) for i in range(1,n+1)]
#Plot the diagonal
line=plt.plot(x, x)
plt.setp(line, linewidth=2, color='r')
w1_min = w1_list[ind]
#Plot the actuals
y = [w1_min*norm.cdf(sorted_ret[j],loc=mu,scale=sigma1)+\
     (1-w1_min)*norm.cdf(sorted_ret[j],loc=mu,scale=sigma2) for j in range(n)]
plt.scatter(x, y, s=5, c='g')
## Configure the graph
plt.title('P-P plot '+lgdates[0][:4]+'-'+lastday[:4]+ \
          ', '+str(n)+' observations')
plt.xlabel('Fraction between 0 and 1')
plt.ylabel('Mixture Normals CDF of Portfolio Log-return')
plt.grid(True)
plt.show()
```
