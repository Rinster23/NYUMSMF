# Q1

## (a) 

```python
import pandas as pd
import numpy as np

data = pd.read_csv('sent_train.csv')

def remove_url(s):
    start = s.find('https')
    if start != -1:
        return s[:start]
    else:
        return s

data['text'] = data['text'].apply(remove_url)
texts = np.array(data['text'])
print(len(texts))
label = data['label'].to_list()
```

## (b)

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer

def doc2vec(doc, tfidf_vectorizer):
    vec = tfidf_vectorizer.transform([doc]).toarray().flatten()
    doc_features = tfidf_vectorizer.get_feature_names_out()
    cnt_vectorizer = CountVectorizer(vocabulary=tfidf_vectorizer.vocabulary_)
    # Fit and transform with CountVectorizer
    doc_counts= cnt_vectorizer.fit_transform([doc]).toarray().flatten()
    return vec, doc_features, doc_counts
```

```python
vectorizer = TfidfVectorizer(
    max_df=0.5, # ignore terms which occur in more than half of the documents
    max_features=10000,
    min_df=2, # ignore terms which occur in less than 2 documents
    stop_words='english',
    norm='l2',
    use_idf=True, 
    analyzer='word',
    # token_pattern='(?u)\\b\\w\\w+\\b'
    token_pattern = '(?u)\\b[a-zA-Z]\\w+\\b'
    )

# note how changing the token_pattern changes the output below
X_tfidf = vectorizer.fit_transform(texts)
print(X_tfidf.shape)
print('first 10 features:', vectorizer.get_feature_names_out()[:10])
print('last 10 features:', vectorizer.get_feature_names_out()[-10:])
```

## (c) 

```python
doc1 = 'Jabberwocky'
doc2 = 'buy $MSFT sell $AAPL hold Brent'
doc3 = 'bullish #stocks'
res1,_,_ = doc2vec(doc1,vectorizer)
res2,_,_ = doc2vec(doc2,vectorizer)
res3,_,_ = doc2vec(doc3,vectorizer)
print('Vector for Jabberwocky: ', res1,sep='\n')
print('Vector for buy $MSFT sell $AAPL hold Brent: ', res2,sep='\n')
print('Vector for bullish #stocks: ', res3,sep='\n')
```

## (d)

```python
from sklearn.decomposition import TruncatedSVD
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import Normalizer
from sklearn.neighbors import KNeighborsClassifier

class myKNN(KNeighborsClassifier):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        
    def transform(self,X,y=None):
        return self
    
    def predict(self, X):
        indices = self.kneighbors(X,return_distance=False)
        res = [texts[indices[i]] for i in range(len(indices))]
        return res


def get_pipeline(n_components):
    svd = TruncatedSVD(n_components=n_components, random_state=24)
    
    lsarec = make_pipeline(
    TfidfVectorizer(
    max_df=0.5, # ignore terms which occur in more than half of the documents
    max_features=10000,
    min_df=2, # ignore terms which occur in less than 2 documents
    stop_words='english',
    norm='l2',
    use_idf=True, 
    analyzer='word',
    # token_pattern='(?u)\\b\\w\\w+\\b'
    token_pattern = '(?u)\\b[a-zA-Z]\\w+\\b'
    ),
    svd, 
    myKNN(n_neighbors=5, algorithm='brute', p=2)
)
    X_lsa = lsarec.fit_transform(texts, label)
    return lsarec
```

## (e)

```python
X = ['Fed', '$MSFT', 'tech rally', 'disappoint earnings']

for n in [50,200,500]:
    my_pipeline = get_pipeline(n)
    res = my_pipeline.predict(X)
    print('#'*40)
    print('Under n_components = ' + str(n)+':')
    for i in range(len(X)):
        print(X[i], res[i],sep='\n')
```

## (f)


```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from nltk import word_tokenize          
from nltk.stem import WordNetLemmatizer
import nltk
from nltk.corpus import stopwords

# Download stopwords list
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
stop_words = set(stopwords.words('english')) 

# Interface lemma tokenizer from nltk with sklearn
class LemmaTokenizer:
    ignore_tokens = [',', '.', ';', ':', '"', '``', "''", '`']
    def __init__(self):
        self.wnl = WordNetLemmatizer()
    def __call__(self, doc):
        return [self.wnl.lemmatize(t) for t in word_tokenize(doc) if t not in self.ignore_tokens]

# Lemmatize the stop words
tokenizer=LemmaTokenizer()
token_stop = tokenizer(' '.join(stop_words))

def get_pipeline_lem(n_components):
    svd = TruncatedSVD(n_components=n_components, random_state=24)
    
    lsarec = make_pipeline(
    TfidfVectorizer(
    max_df=0.5, # ignore terms which occur in more than half of the documents
    max_features=10000,
    min_df=2, # ignore terms which occur in less than 2 documents
    stop_words=token_stop,
    tokenizer=tokenizer,
    norm='l2',
    use_idf=True, 
    analyzer='word',
    # token_pattern='(?u)\\b\\w\\w+\\b'
    token_pattern = '(?u)\\b[a-zA-Z]\\w+\\b'
    ),
    svd, 
    myKNN(n_neighbors=5, algorithm='brute', p=2)
)
    X_lsa = lsarec.fit_transform(texts, label)
    return lsarec
```

```python
for n in [50,200,500]:
    my_pipeline = get_pipeline_lem(n)
    res = my_pipeline.predict(X)
    print('#'*40)
    print('Under n_components = ' + str(n)+':')
    for i in range(len(X)):
        print(X[i], res[i],sep='\n')
```

### We can see that the most results changed when we lemmatize before the TFIDF step. Since we handled Morphological Variations by lemmatization, which reduced words to a common base form, the new results are more meaningful and richer in content.

## (a) 

```python
import numpy as np
from sklearn.base import BaseEstimator, TransformerMixin
import matplotlib.pyplot as plt

class GaussianKernelFeatures(BaseEstimator, TransformerMixin):
    def __init__(self, locations, sigma=1.0):
        self.locations = locations
        self.sigma = sigma

    def fit(self, X, y=None):
        # No fitting necessary, but we implement the method to comply with scikit-learn's TransformerMixin
        return self

    def transform(self, X):
        # Ensure X is a 2D column vector
        X = np.array(X).reshape(-1, 1)
        
        # Calculate the Gaussian kernel feature map
        features = np.zeros((X.shape[0], len(self.locations)))
        for i, x0 in enumerate(self.locations):
            features[:, i] = np.exp(-(X - x0) ** 2 / (2 * self.sigma ** 2)).squeeze()
            
        # Normalize by the constant factor
        features /= np.sqrt(2 * np.pi * self.sigma ** 2)
        return features
```

## (b) 

```python

locations = np.linspace(-2, 2, 10)  # Define 10 locations for the centers of the Gaussian kernels
gkf = GaussianKernelFeatures(locations=locations, sigma=0.1)

x_predict = np.linspace(-2, 2, 1000)
features = gkf.fit_transform(x_predict)  # Apply the Gaussian kernel feature mapping


plt.figure(figsize=(12,8))
plt.plot(x_predict, features,'o')
plt.show()
```

## (c)

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, RANSACRegressor, TheilSenRegressor, HuberRegressor
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error

np.random.seed(42)
X = np.random.uniform(low=-2, high=2, size=500)

model = 'periodic'

if model == 'polynomial':
    y = 2. + 2 *X  + .4*X**2 - .4 * X**3
elif model == 'periodic':
    y = np.sin(20 * np.pi * 0.1 * X)
    
X = X[:,np.newaxis]

# random normal error
error = 0.2 * np.random.randn(y.shape[0])

# generate observations with additive noise
y += error 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)



gkf = GaussianKernelFeatures(locations, 0.55)

estimators = [('Least-Square', '-', 'C0', LinearRegression(fit_intercept=False)),
              ('Theil-Sen', '-', 'C1', TheilSenRegressor(random_state=42)),
              ('RANSAC', '-', 'C2', RANSACRegressor(random_state=42)),
              ('HuberRegressor', '-', 'C3', HuberRegressor())]

fig, ax = plt.subplots(1, 1, figsize=(16, 10))
fig.suptitle('Gaussian Kernel Regression')
ax.plot(X_train, y_train, 'o', ms=5, c='black',
        label='data points [10% outliers]')

for label, style, color, estimator in estimators:
    model = make_pipeline(gkf, estimator)
    model.fit(X_train, y_train)
    mse = mean_squared_error(model.predict(X_test), y_test)
    y_predicted = model.predict(x_predict[:, None])
    ix = np.argsort(x_predict)
    ax.plot(x_predict[ix], y_predicted[ix], style, lw=2,
            color=color, label=label + ' E={:2.2g}'.format(mse))
ax.legend(loc='upper right', framealpha=0.95)
# ax.set(ylim=(-2, 8), xlabel='time [s]', ylabel='amplitude')
plt.show()
```

### Best parameter searching for periodic data:

```python
from sklearn.model_selection import GridSearchCV 

param_grid = {'gaussiankernelfeatures__sigma': [0.1+0.05*i for i in range(0, 19)]}
estimator = LinearRegression(fit_intercept=False)

model = make_pipeline(gkf, estimator)

grid_search = GridSearchCV(model, param_grid, cv=3, scoring='neg_mean_squared_error', verbose=1)

# Fit the model with grid search
grid_search.fit(X, y)

# Best parameters and model
best_params = grid_search.best_params_
best_mse = grid_search.best_score_

print('Best sigma: ', best_params)
print('Best MSE: ', -best_mse)
```

## (d)

```python
bm = [0]
X = np.linspace(0,4,100)
gap = X[1] - X[0]
coef = np.sqrt(gap)
np.random.seed(42) 
z = np.random.normal(0,1,99)
for i in range(99):
    next_val = bm[-1] + coef*z[i]
    bm.append(next_val)
```

```python
plt.plot(X,bm)
plt.show()
```

```python
X = X[:,np.newaxis]
x_pred = np.linspace(0,4,1000)

for k in [5,20,50,100]:
    locations = np.linspace(0,4,k)
    gkf = GaussianKernelFeatures(locations, 1)
    
    param_grid = {'gaussiankernelfeatures__sigma': [0.1+0.1*i for i in range(0, 50)]}
    estimator = LinearRegression(fit_intercept=False)
    model = make_pipeline(gkf, estimator)
    grid_search = GridSearchCV(model, param_grid, cv=3, scoring='neg_mean_squared_error', verbose=1)

    # Fit the model with grid search
    grid_search.fit(X, bm)
    print(X.shape)
    # Best parameters and model
    best_params = grid_search.best_params_
    print('For K=' +str(k)+':')
    print('Best sigma: ', best_params)
    fig, ax = plt.subplots(1, 1, figsize=(16, 10))
    ax.plot(X, bm, 'o', ms=5, c='black')
    
    gkf = GaussianKernelFeatures(locations, best_params['gaussiankernelfeatures__sigma'])
    model = make_pipeline(gkf, estimator)
    model.fit(X, bm)
    mse = mean_squared_error(model.predict(X), bm)
    y_predicted = model.predict(x_pred[:, None])
    ax.plot(x_pred, y_predicted, '-', lw=2, color='b')
    fig.suptitle('Gaussian Kernel Regression for K={}, sigma={}, mse={}'\
                 .format(k,round(best_params['gaussiankernelfeatures__sigma'],2),round(mse,2)))
    plt.show()
```

### We can see that as $K$ increases, the MSE decreases but will not keep decreasing. This is because that when we increase the number of independent variables from 5 to 20 in linear regression, the fitting can be imporved; but if we increase from 20 to 100, the effect may not be that significant.
