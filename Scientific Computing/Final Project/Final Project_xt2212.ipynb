## The paper I chose is:

## Merton, R.C. (1976) Option Pricing when Underlying Stock Returns are Discontinuous

## Brief Intro

The Black-Scholes model is one of the classical models of option pricing. However, the BS model
is not perfect. It is based on several strict assumptions. For example, interest rate and volatility
are known constants, it is ignoring dividends and transaction costs; stock index prices walk
randomly in a continuous time series and the return follows a log normal distribution. This is
inconsistent with the real financial data, which has the characteristics of leptokurtic. Due to
industry shocks, policy shocks, financial problems, and other situations, stock prices do not walk
randomly in the continuous time series but have obvious jump characteristics. At the same time,
the BS model is also insufficient to estimate the “tail risk”. The BS model assumes that the
volatility of stock prices is constant, but in fact, the volatility changes over time. In addition,
the BS model assumes that stock prices follow a log normal distribution. However, in reality, the
distribution of stock prices may deviate from the log normal distribution due to a wide variety
of factors. These imply that BS models cannot consistently predict option prices accurately. Merton jump diffusion (we call it MJD) model is one of the first beyond Black-Scholes model in the sense that it tries to capture the discontinuity of stock prices commonly observed in the market. By adding a compound Possion jump process to the geometric brownian motion, MJD model captures the negative skewness and excess kurtosis of the log stock price density, and the three extra parameters $\lambda,\mu,\delta $ allows users to control skewness and excess kurtosis. Under this model, closed-form pricing formulas for European call option can also be derived.

## Model Derivation

Merton assumes that the jump sizes follow log-normal distribution, i.e., the jump (if any) within a small time interval is $\frac{dS_t}{S_t} = \frac{y_tS_t - S_t}{S_t} = y_t-1$, where $\ln(y_t)\sim N(\mu, \delta^2)$, and $y_t-1$ has mean $\exp\{\mu+\frac{1}{2}\delta^2\}-1\equiv k$. Denote $Y_t = \ln(y_t)\sim i.i.d\ N(\mu, \delta^2)$.

MJD dynamics of asset price which incorporates the above properties takes the SDE of the form:
$$
\frac{dS_t}{S_t} = (\alpha-\lambda k)dt+\sigma dB_t + (y_t-1)dN_t
$$

where $\alpha$ is the instantaneous expected return on the asset, $\sigma$ is the instantaneous volatility of the asset return conditional on that jump does not occur, $B_t$ is a standard Brownian motion process, and $N_t$ is an Poisson process with intensity $\lambda$. Standard assumption is that $B_t, N_t$ and $y_t$ are independent.

The expected relative price change from the jump part $dN_t$ in the time
interval $dt$ is $\lambda k dt$. This is the predictable part of the jump. This is why the instantaneous expected return on the asset $\alpha dt$ is adjusted by $-\lambda k dt$ in the drift term of the jump-diffusion process to make the jump part an unpredictable innovation.

We can solve this SDE using Itô formula for jump processes:
$$
S_t = S_0\exp\{\alpha - \frac{\sigma^2}{2} - \lambda k)t+\sigma B_t+\sum_{i=1}^{N_t}Y_i\}
$$

Next we derive the closed form pricing formula for MJD model using martingale approach.

## Pricing Formula

MJD model is an example of an incomplete model because there are many equivalent martingale risk-neutral measures $\mathbb{Q}\sim\mathbb{P}$ under which the discounted stock price is a martingale. Merton proposes his risk neutral measure by changing the drift of the Brownian motion process while keeping the other parts **(most important is the jump measure, i.e. the distribution of jump times and jump sizes) unchanged**:

$$
S_t = S_0\exp\{(r-\frac{\sigma^2}{2}-\lambda k)t+\sigma B_t^\mathbb{Q}+\sum_{k=1}^{N_t}Y_k \}
$$

where $B_t^{\mathbb{Q}}$ is a standard Brownian motion and the process $\{e^{-rt}S_t;0\leq t\leq T\}$ is a martingale under $\mathbb{Q}$, $N_t$ is still a poisson process with intensity $\lambda$ and $Y_k\sim N(\mu, \delta^2)$. Then the European option price with payoff function H(ST ) is calculated as


\begin{aligned}
V(t,S_t) &= e^{-r(T-t)} \mathbb{E}^{\mathbb{Q}} [H(S_T)|\mathcal{F}_t] \\
& = e^{-r\tau}\mathbb{E}^{\mathbb{Q}} \left[H\left(S_t\exp\left\{(r-\frac{\sigma^2}{2}-\lambda k )\tau+\sigma B_{\tau}^\mathbb{Q}+\sum_{k=1}^{N_t}Y_k \right\}\right)|S_t\right] \\
&= e^{-r\tau}\mathbb{E}^{\mathbb{Q}} \left[H\left(S_t\exp\left\{(r-\frac{\sigma^2}{2}-\lambda k )\tau+\sigma B_{\tau}^\mathbb{Q}+\sum_{k=1}^{N_t}Y_k \right\}\right)\right]
\end{aligned}

where $\tau = T-t$.

Using tower law, we can write
\begin{aligned}
V(t,S_t) &= e^{-r\tau}\mathbb{E}^{\mathbb{Q}}\left\{\mathbb{E}^{\mathbb{Q}} \left[H\left(S_t\exp\left\{(r-\frac{\sigma^2}{2}-\lambda k )\tau+\sigma B_{\tau}^\mathbb{Q}+\sum_{k=1}^{N_t}Y_k \right\}\right) \Bigg|N_t \right]\right\} \\
&= e^{-r\tau} \sum_{i=0}^{\infty}\pi_i\mathbb{E}^{\mathbb{Q}} \left[H\left(S_t\exp\left\{(r-\frac{\sigma^2}{2}-\lambda k )\tau+\sigma B_{\tau}^\mathbb{Q}+\sum_{k=1}^{i}Y_k \right\}  \right)\Bigg|N_t=i \right]  \\
&= e^{-r\tau}\sum_{i=0}^{\infty}\pi_i\mathbb{E}^{\mathbb{Q}} \left[H\left(S_t\exp\left\{(r-\frac{\sigma^2}{2}-\lambda k )\tau+\sigma B_{\tau}^\mathbb{Q}+\sum_{k=1}^{i}Y_k \right\} \right)\right]
\end{aligned}

where $\pi_i = \frac{e^{-\lambda \tau}(\lambda \tau)^i}{i!}$.

$$
(r-\frac{\sigma^2}{2}-\lambda k )\tau+\sigma B_{\tau}^\mathbb{Q}+\sum_{k=1}^{i}Y_k \sim N((r-\frac{\sigma^2}{2}-\lambda k )\tau+i\mu,\sigma^2\tau+i\delta^2)
$$

We can rearrange the terms such that the distribution remains the same:

$$
(r-\frac{\sigma^2}{2}-\lambda k )\tau+i\mu +\sqrt{\frac{\sigma^2\tau+i\delta^2}{\tau}}B_{\tau}^\mathbb{Q}\sim N((r-\frac{\sigma^2}{2}-\lambda k )\tau+i\mu,\sigma^2\tau+i\delta^2)
$$

Substitute $k = \exp\{\mu+\frac{1}{2}\delta^2\}-1$, we have

\begin{aligned}
V(t,S_t) &= e^{-r\tau}\sum_{i=0}^{\infty}\pi_i\mathbb{E}^{\mathbb{Q}} \left[H\left(S_t\exp\left\{(r-\frac{\sigma^2}{2}-\lambda (e^{\mu+\frac{1}{2}\delta^2}-1) )\tau+ i\mu +\sqrt{\frac{\sigma^2\tau+i\delta^2}{\tau}}B_{\tau}^\mathbb{Q}\right\} \right)\right]
\end{aligned}

\begin{aligned}
V(t,S_t) &= e^{-r\tau}\sum_{i=0}^{\infty}\pi_i\mathbb{E}^{\mathbb{Q}} \left[H\left(S_t\exp\left\{(r-\frac{\sigma^2}{2}+(\frac{i\delta^2}{2\tau} - \frac{i\delta^2}{2\tau})-\lambda (e^{\mu+\frac{1}{2}\delta^2}-1) )\tau+ i\mu +\sqrt{\frac{\sigma^2\tau+i\delta^2}{\tau}}B_{\tau}^\mathbb{Q}\right\} \right)\right] \\
&= e^{-r\tau} \sum_{i \geq 0} \frac{e^{-\lambda \tau} (\lambda\tau)^i}{i!} \mathbb{E}^{\mathbb{Q}} \left[ H \left( S_t \exp \left( i \mu + \frac{i \delta^2}{2} - \lambda(e^{\mu+\delta^2/2} - 1)\tau \right) \exp \left( ( r - \frac{1}{2} \sigma_i^2)\tau + \sigma_i B_{\tau}^\mathbb{Q}\right) \right) \right]
\end{aligned}

where $\sigma_i^2 = \sigma^2 + \frac{i\delta^2}{\tau}$

Note that in BS, we have

$$
V^{BS}(\tau = T - t, S, \sigma) = e^{-r\tau} \mathbb{E}^{\mathbb{Q_{BS}}} \left[ H\left(S_t\exp\left( (r - \frac{\sigma^2}{2}) \tau + \sigma B_{\tau}^{\mathbb{Q}_{BS}} \right) \right) \right]
$$


Finally, MJD pricing formula can be obtained as a weighted average of Black-Scholes price conditioned on the number of jumps $i$ :

$$
V^{\text{Merton}}(t, S_t) = \sum_{i \geq 0} \frac{e^{-\lambda \tau} (\lambda \tau)^i}{i!} V^{BS}\left(\tau, S_i\equiv S_t\exp \left( i \mu + \frac{i \delta^2}{2} - \lambda(e^{\mu+\delta^2/2} - 1)\tau \right); \sigma_i \equiv \sqrt{ \sigma^2 + \frac{i\delta^2}{\tau}}\right)
$$


## Empirical Analysis 

Next we will carry out parameter estimation to calibrate the MJD model against real-world data of options on Standard and Poor’s 500 index (S&P 500). We will also compare the error generated by Black-Scholes formula and MDJ model.

```python
import numpy as np
import pandas as pd
import math
from scipy.stats import norm
from scipy.optimize import minimize_scalar, minimize
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
N = norm.cdf

# pricing formula for BS model 
def BS_CALL(S, K, T, r, sigma):
    d1 = (np.log(S/K) + (r + sigma**2/2)*T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    return S * N(d1) - K * np.exp(-r*T)* N(d2)

# pricing formula for MJD model
def MJD_CALL(S, K, T, r, params, n_terms=20):
    p = 0
    sigma,mu,delta,lam = params
    for i in range(n_terms):
        S_i = S * np.exp(i*mu+i*delta**2/2-lam*(np.exp(mu+delta**2/2)-1)*T)
        sigma_i = np.sqrt(sigma**2+i*delta**2/T)
        pi_i = np.exp(-lam*T)*(lam*T)**i / math.factorial(i)
        p +=  pi_i * BS_CALL(S_i, K, T, r, sigma_i)
    return p 
```

Next we will calibrate the models against market price data fixing time to maturity. We get the data from https://www.cboe.com/delayed_quotes/spx/quote_table.

We used time to maturity = 7, 24, 59 and 87 days, all European call options. The prices are observed after the market closes on 2024-04-23, and the undersying ends with 5070.55.

```python
from sklearn.base import BaseEstimator, ClassifierMixin, TransformerMixin
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error
from tqdm.notebook import tqdm
import warnings
warnings.filterwarnings("ignore")

class CalibrateBS(BaseEstimator, TransformerMixin):
    def __init__(self, S, T, r):
        """
        Initialize the CalibrateBS class with required parameters.
        
        Parameters:
        S (float): Current asset price.
        T (float): Time to maturity.
        r (float): Risk-free interest rate.
        """
        super().__init__()
        self.S = S
        self.T = T
        self.r = r

    def fit(self, Ks, y):
        """
        Fit method to calibrate the Black-Scholes model.

        Parameters:
        Ks (pandas Series or DataFrame): Strike prices.
        y (pandas Series or DataFrame): Observed option prices.

        """
        Ks = Ks.values
        y = y.values
        
        # Define objective function to minimize
        def obj(sigma):
            err = 0
            for i in range(len(Ks)):
                # Calculate squared error for each option
                err += (BS_CALL(self.S, Ks[i], self.T, self.r, sigma) - y[i]) ** 2
            return err
        
        # Minimize the objective function to find optimal sigma
        res = minimize_scalar(obj, bounds=(0, 5))
        self.sigma = res.x
        
    def predict(self,Ks,y=None):
        """
        Predict method to calculate option prices using the calibrated Black-Scholes model.

        Parameters:
        Ks (pandas Series or DataFrame): Strike prices.
        y (pandas Series or DataFrame): Observed option prices (optional, not used in prediction).

        Returns:
        list: Predicted option prices.
        """
        Ks =  Ks.values
        res = []
        for k in Ks:
            res.append(BS_CALL(self.S, k, self.T, self.r, self.sigma))
        return res
        


class CalibrateMJD(BaseEstimator, TransformerMixin):
    def __init__(self, S, T, r, initial_guess=None):
        """
        Initialize the CalibrateMJD class with required parameters.
        
        Parameters:
        S (float): Current asset price.
        T (float): Time to maturity.
        r (float): Risk-free interest rate.
        initial_guess (array-like, optional): Initial guess for parameter optimization.
        """
        super().__init__()
        self.S = S
        self.T = T
        self.r = r
        self.initial_guess = initial_guess
    
    def fit(self, Ks, y):
        """
        Fit method to calibrate the Merton Jump Diffusion (MJD) model.

        Parameters:
        Ks (pandas Series or DataFrame): Strike prices.
        y (pandas Series or DataFrame): Observed option prices.
        """
        Ks = Ks.values
        y = y.values 
        
        # Define parameter bounds
        bnds = ((0,None),(None,None),(0,None),(0,None))
        
        # Define objective function to minimize
        def obj(params):
            err = 0 
            for i in range(len(Ks)):
                # Calculate squared error for each option
                err += (MJD_CALL(self.S,Ks[i],self.T,self.r,params) - y[i])**2
            return err
        
        # Perform optimization with or without initial guess
        if self.initial_guess:
            res = minimize(obj,self.initial_guess,bounds=bnds,method='Nelder-Mead')
        else:
#             # for optimization methods other than Nelder-Mead, we may need to sample initial guess many times 
#             # to get a good calibration
#             # sample initial guess
#             params_arr = []
#             diffs = []
#             N = 10
#             for j in tqdm(range(N),desc='Trying initial guess'):
#                 x0 = [np.random.uniform(0.01, 0.5),np.random.uniform(-1,1), np.random.uniform(0,2),np.random.randint(1, 5)]
#                 res = minimize(obj, x0, method='Nelder-Mead', bounds=bnds)
#                 params_arr.append(res.x)
#                 diffs.append(res.fun)
#             raw_params = params_arr[np.argmin(diffs)]
            # Generate random initial parameters within specified ranges
            raw_params = [np.random.uniform(0.01, 0.5),np.random.uniform(-1,1), np.random.uniform(0,2),np.random.randint(1, 5)]
            res = minimize(obj, raw_params, method='Nelder-Mead', bounds=bnds)
        self.params = res.x

    def predict(self,Ks,y=None):
        Ks =  Ks.values
        res = []
        for k in Ks:
            res.append(MJD_CALL(self.S, k, self.T, self.r, self.params))
        return res
```

```python
def model_comparison(sheet_name, tenor, initial_guess=None): # sigma,mu,delta,lam
    """
    Compare Black-Scholes (BS) and Merton Jump Diffusion (MJD) models using option data.

    Parameters:
    sheet_name (str): Name of the sheet in the Excel file containing option data.
    tenor (float): Time to maturity.
    initial_guess (array-like, optional): Initial guess for MJD model parameter optimization.

    """
    data = pd.read_excel('option_data.xlsx',sheet_name=sheet_name)
    Ks = data['Strike']
    y = (data['Ask']+data['Bid'])/2
    # Calibrate BS model
    BS = CalibrateBS(S,tenor,r)
    BS.fit(Ks, y)
    # Calibrate MJD model
    MJD = CalibrateMJD(S,tenor,r,initial_guess)
    MJD.fit(Ks, y)
    # Predict option prices using calibrated models
    fitted_BS = BS.predict(Ks)
    fitted_MJD = MJD.predict(Ks)
    print(f'##################### T = {round(tenor,4)} #####################')
    # Plot comparison of market prices and fitted prices
    plt.figure(figsize=(10,8))
    plt.plot(data['Strike'],y, label='Market price')
    plt.plot(Ks, fitted_BS, label='Fitted price in BS')
    plt.plot(Ks, fitted_MJD, label='Fitted price in MDJ')
    plt.legend()
    plt.xlabel('Strike')
    plt.ylabel('Option price')
    plt.title(f'Model comparison for T = {round(tenor,4)}')
    plt.show()
    # Calculate 5-fold Cross Validation scores
    BS_CV_score = np.mean(cross_val_score(CalibrateBS(S,tenor,r),Ks,y,scoring='neg_mean_squared_error',cv=5))
    MJD_CV_score = np.mean(cross_val_score(CalibrateMJD(S,tenor,r,list(MJD.params)),Ks,y,scoring='neg_mean_squared_error',cv=5))
    print('5 fold Cross Validation score of BS model, ', -BS_CV_score)
    print('5 fold Cross Validation score of MJD model, ', -MJD_CV_score)
```

We can see that Merton's jump diffusion model greatly outperforms BS formula on the 5 fold cross validation. It is not quite surprising since there are four free parameters in Merton's model, hence, it can take more market dynamics into the model, which should provide better results.

However, more paramters also means more computing resource during model calibration. It takes much longer time to calibrate MJD model than BS model. Besides, since there are four free parameters and their bounds are quite flexible, i.e., $(0,\infty)$ for $\sigma$, no constraints on $\mu$, $(0,\infty)$ for $\delta$, $(0,\infty)$ for $\lambda$. Hence, choosing a proper optimization method is very important. In the first place I chosed SLSQP as the optimization method, and I found it very sensitive to the initial guess. Therefore, I generated 100 random initial guesses and chose the top one as the final initial guess. Later I tried Nelder-Mead optimization method, which is much better in this problem and not sensitive to the initial guess.

In conclusion, Merton's jump diffusion model is a more sophisticated option pricing model, and to some extent can provide better calibration to the market option price data.

**Additional Study: Differences Between SLSQP and Nelder-Mead Optimization Method**

In the Empirical Analysis part, we found that SLSQP (Sequential Least Squares Quadratic Programming) and Nelder-Mead are both optimization algorithms used to find the minimum (or maximum) of a function. We observed that Nelder-Mead optimization method is less sensitive to the initial guess. The following is a breakdown of the main differences between the two:

**Type of Optimization**：
- SLSQP: SLSQP is a gradient-based optimization method. It requires the gradient (or derivative) of the objective function to be minimized.
- Nelder-Mead: Nelder-Mead is a derivative-free optimization method. It does not require knowledge of the gradient of the objective function.

**Convergence Behavior**:

- SLSQP: As a gradient-based method, SLSQP generally converges faster when the objective function is smooth and well-behaved.
- Nelder-Mead: Nelder-Mead may be more robust in handling non-smooth or discontinuous objective functions because it does not rely on gradient information. However, it may converge slower than gradient-based methods in some cases.

**Initialization Sensitivity**:
- SLSQP: SLSQP can be sensitive to the initial guess or starting point, especially if the objective function has multiple local minima or is highly nonlinear.
- Nelder-Mead: Nelder-Mead is less sensitive to the initial guess and often requires less tuning of parameters.

**Memory and Computational Requirements**:
- SLSQP: SLSQP typically requires more memory and computational resources, especially for large-scale optimization problems, due to the need to compute and store gradients.
- Nelder-Mead: Nelder-Mead is often more memory-efficient and computationally simpler since it does not involve gradient computations.
