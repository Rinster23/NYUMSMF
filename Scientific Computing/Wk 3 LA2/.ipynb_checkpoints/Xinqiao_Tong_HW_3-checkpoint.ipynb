```python
import random
import pandas as pd
import numpy as np
import numpy.linalg as la
import yfinance as yf
import warnings
warnings.simplefilter("ignore")
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
rtns = pd.read_pickle('/content/drive/My Drive/Colab Notebooks/SP500_rtn_data.pkl')
std_rtns = rtns.copy(deep=True)
col_names = list(std_rtns.columns)
std_rtns[col_names] = StandardScaler().fit_transform(std_rtns[col_names])

def lamplus(n,r):
    # Upper eigenvalue limit of the Marchenko Pastur density
    q = float(n)/float(r)
    return (1+1/q+2*np.sqrt(1/q))

eig_val_temp = np.zeros(std_rtns.shape[1])
for i in range(1000):
    # Shuffle every column in the DataFrame
    shuffled_rtns = std_rtns.apply(np.random.permutation,axis=0)
    # Compute correlation matrix
    correlation_matrix = shuffled_rtns.corr()
    # Compute eigenvalue/vector decomposition of matrix
    eigenvalues, _ = la.eig(correlation_matrix)
    eig_val_temp += eigenvalues

eig_val_temp /= 1000
```

```python
t,m = std_rtns.shape
filtvals = [val for val in eigenvalues if val < lamplus(t,m)]     # filter eigenvalues
sevlist = [np.mean(filtvals)] * len(filtvals)
feval = eigenvalues[:(len(eigenvalues)-len(sevlist))].tolist() + sevlist
print('Number of eigenvalues larger than lambda^+: ',len(eigenvalues) - len(filtvals))

def eigden(lam,n,r):
    '''
    Definition of the Marchenko Pastur density
    '''
    q = float(n)/r
    lplus = 1+1/q+2*np.sqrt(1/q)
    lminus = 1+1/q-2*np.sqrt(1/q)
    soln = q/(2*np.pi*lam)*np.sqrt((lplus-lam)*(lam-lminus))
    return soln

lamvls = np.linspace(0.001,3,2000)
plt.figure(figsize=(14,5))
plt.hist([val for val in eigenvalues],bins=75, edgecolor='black', density=True);
plt.plot(lamvls,[eigden(lam,len(std_rtns),len(std_rtns.columns)) for lam in lamvls], color="r", linewidth=3);
plt.xlabel("Eigenvalue Bins",fontsize=16)
plt.ylabel("Probability Mass",fontsize=16)
plt.title("Histogram of Mean Eigenvalues from Correlation Matrix of Randomized S&P Returns and Marcenko Pastur Density", fontsize=20)
plt.show()
```

#### We can see that when we shuffle the returns in each time series independently and take average of the eigenvalues, none of the eigenvalues are larger than than the upper bound of Marcenko Pastur range, which indicates that all the eigenvalues are noise eigenvalues. This is true in fact since we randomized returns are noise themselves. Besides, since here our correlation matrix is generated from noise, the plot also justifies Marchenko Pastur density theorem for eigenvalue spectrum of random correlation matrix as the range of our eigenvalue fits with the theoretical range very well.

#### (2)

Direction of Maximum Variance: Each eigenvector represents a direction in the original feature space. The direction with the largest eigenvalue corresponds to the direction of maximum variance in the dataset.

Orthogonality: Eigenvectors are orthogonal to each other, meaning they are perpendicular to each other in the feature space. This implies that the principal components capture different directions of variation in the data.

Ranking of Components: Eigenvectors are typically sorted based on their corresponding eigenvalues, with the eigenvector corresponding to the largest eigenvalue representing the first principal component, the second largest eigenvalue representing the second principal component, and so on.

Feature Contribution: The components of an eigenvector represent the contribution of each original feature to that principal component. Larger absolute values indicate stronger contributions from the corresponding features.

Dimensionality Reduction: Eigenvectors are used to transform the original data into a new, lower-dimensional space where each dimension (principal component) captures a significant portion of the variance in the dataset.

Visualization: Eigenvectors can be visualized as vectors in the original feature space or as axes in the transformed space. Visualizing the data projected onto the principal components can help understand the relationship between features and the variance captured by each component.

```python
# (3)
print('Eigenvalues based on covariance: \n', DJI_eigenvalues)
print('Eigenvalues based on correlation: \n', DJI_eigenvalues2)
bar_width = 0.3
indices = np.arange(len(DJI_eigenvalues))
plt.figure(figsize=(14,5))
plt.bar(indices - bar_width/2, sorted(DJI_eigenvalues,reverse=True), bar_width, color='b', label='Eigenvalues from Cov')
plt.bar(indices + bar_width/2, sorted(DJI_eigenvalues2,reverse=True), bar_width, color='r', label='Eigenvalues from Corr')
plt.xlabel('Eigenvalues')
plt.title('Plot of eigenvalues')
plt.legend()
plt.show()
```

#### We can see that eigenvalues based on covariance are larger than those based on correlation, perhaps because correlation matrix is the covariance matrix of the standardized return, which reduces the magnitude of data.

```python
# (4)
import pandas as pd
import numpy as np
import yfinance as yf
from datetime import datetime, timedelta

df = pd.read_html("https://en.wikipedia.org/wiki/Dow_Jones_Industrial_Average")[1]
tickers = df['Symbol'].to_list()
end_date = datetime.today()
data = yf.download(tickers, start='2021-01-04', end=end_date.strftime('%Y-%m-%d'), auto_adjust=True)['Close'].pct_change()
# Clean data so there are continuous time series for each ticker
clean1 = data.iloc[1: , :] # Drop first row of NaNs
clean2 = clean1.dropna(axis=1) # Drop columns with NaNs
DJI_rnt_full = clean2 * 100
DJI_rnt_full
```

```python
DJI_cov_matrix_full = DJI_rnt_full.cov()
DJI_eigenvalues_full, DJI_eigenvectors_full = la.eig(DJI_cov_matrix_full)
DJI_corr_matrix_full = DJI_rnt_full.corr()
DJI_eigenvalues_full2, DJI_eigenvectors_full2 = la.eig(DJI_corr_matrix_full)
print('Eigenvalues based on covariance from pickle: \n', DJI_eigenvalues)
print('Eigenvalues based on covariance of data till today: \n', DJI_eigenvalues_full)
print('Eigenvalues based on correlation from pickle: \n', DJI_eigenvalues2)
print('Eigenvalues based on correlation of data till today: \n', DJI_eigenvalues_full2)

plt.bar(indices - bar_width/2, sorted(DJI_eigenvalues,reverse=True), bar_width, color='b', label='Eigenvalues from Cov from pickle')
plt.bar(indices + bar_width/2, sorted(DJI_eigenvalues_full,reverse=True), bar_width, color='r', label='Eigenvalues from Cov till now')
plt.xlabel('Eigenvalues')
plt.title('Plot of eigenvalues')
plt.legend()
plt.show()
plt.bar(indices - bar_width/2, sorted(DJI_eigenvalues2,reverse=True), bar_width, color='b', label='Eigenvalues from Corr from pickle')
plt.bar(indices + bar_width/2, sorted(DJI_eigenvalues_full2,reverse=True), bar_width, color='r', label='Eigenvalues from Corr till now')
plt.xlabel('Eigenvalues')
plt.title('Plot of eigenvalues')
plt.legend()
plt.show()
```

#### We can see that most eigenvalues of the covariance matrix become larger when we extend the time period till today, while those from the correlation matrix become smaller. This suggests that the methods like PCA which based on eigendecomposition are not necessarily consistent over time: if you decompose returns on assets for one time period and then decompose the returns on the same set of assets for a different time period, the principal components are not necessarily the same. They could switch order, or they could be entirely different!

#### Besides, as we can see in the plot, since eigen values from covariance matrix are larger than those from correlation matrix, then there would be more eigenvalues larger than lambda^+.

```python
# (5)
ticker_symbol = '^DJI'
end_date = datetime.today().strftime('%Y-%m-%d')
start_date = (datetime.today() - timedelta(days=48)).strftime('%Y-%m-%d')
dji_data = yf.download(ticker_symbol, start=start_date, end=end_date)['Close'].pct_change()
dji_data =  dji_data.iloc[1:]
dji_data *= 100
```

```python
# choose number of eigenvalues such that we can explain 95% variance
total = sum(DJI_eigenvalues_full)
sort_eigval = sorted(zip(DJI_eigenvalues_full, DJI_eigenvectors_full),reverse=True)
cur_sum = 0
i = 0
weight = np.zeros(len(DJI_eigenvalues_full))
while cur_sum/total < 0.95:
    cur_sum += sort_eigval[i][0]
    weight += sort_eigval[i][1]
    i += 1

rep_potf = weight @ (DJI_rnt_full.values.T)
df = pd.DataFrame({'DJI_return':dji_data.values,'Replication_return':rep_potf[-30:]},index=dji_data.index)
plt = df.plot()
```
