```python
import pandas as pd
import numpy as np
import yfinance as yf
from datetime import datetime
from dateutil.relativedelta import relativedelta

df = pd.read_html("https://en.wikipedia.org/wiki/Dow_Jones_Industrial_Average")[1]
tickers = df['Symbol'].to_list()
data = yf.download(tickers, start='2020-01-01', end='2024-02-29', auto_adjust=True)['Close'].pct_change()
# Clean data so there are continuous time series for each ticker
DJIA = data.iloc[1: , :] # Drop first row of NaNs
```

```python
DJIA.head()
```

```python
ans = pd.DataFrame(index=pd.date_range(start = '2020-06-30', end = '2023-12-31', freq = 'M'))
```

```python
from scipy.optimize import minimize

def portf_obj(x,V,mu):
    return x.T @ V @ x - mu.T @ x # also correct without .T

# constraints & bounds
constraints = ({'type': 'eq', 
                'fun': lambda x: np.sum(x) - 1, # weights sum = 1
                'jac': lambda x: np.ones(len(x))})

bounds = [(-1,1) for _ in range(len(DJIA.columns))] # allow short positions

mean_hist = []
risk_hist = []
holdings_hist = []
start_date = datetime.strptime("2020-01-01", "%Y-%m-%d")
initial_weights = np.ones(len(DJIA.columns)) * 1/len(DJIA.columns)

for i in range(len(ans)):
    end_date = start_date + relativedelta(months=6) - relativedelta(days=1)
    data = DJIA.loc[start_date:end_date]
    
    V = data.cov()
    mu = data.mean()
    opt_portf = minimize(fun = portf_obj, # minimize obj fun
                               x0 = initial_weights,
                               args = (V,mu),
                               method='SLSQP',
                               bounds=bounds,
                               constraints=constraints)
    opt_weights = opt_portf.x
    holdings_hist.append(opt_weights)
    mean_hist.append(mu.T @ opt_weights)
    risk_hist.append(opt_weights.T @ V @ opt_weights)
    start_date = start_date + relativedelta(months=1)
```

```python
ans['Expected return'] = mean_hist
ans['Expected risk'] = np.sqrt(risk_hist) # get std
ans['Ret/Risk'] = ans['Expected return']/ans['Expected risk']
ans['Annualized Sharpe'] = ans['Ret/Risk'] * np.sqrt(12)
ans.tail()
```

```python
import matplotlib.pyplot as plt

x_labels = [dt.strftime('%Y-%m') for dt in ans.index]
plt.figure(figsize=(12,5))
plt.plot(np.array(mean_hist)*100)
plt.ylabel('Value %')
plt.xticks(ticks=range(len(mean_hist)),labels=x_labels,rotation=90)
plt.title('Expected Monthly Returns')
plt.show()
plt.figure(figsize=(12,5))
plt.plot(np.sqrt(risk_hist)*100)
plt.ylabel('Value %')
plt.xticks(ticks=range(len(mean_hist)),labels=x_labels,rotation=90)
plt.title('Expected Monthly Risks')
plt.show()
plt.figure(figsize=(12,5))
plt.plot(ans['Ret/Risk'].values)
plt.ylabel('Value')
plt.xticks(ticks=range(len(mean_hist)),labels=x_labels,rotation=90)
plt.title('Monthly Return / Risk')
plt.show()
```

## Problem 2 (50 points):

Often portfolios are constructed where they are intended to “track” another portfolio or index. On the buy side, this can be for performance purposes – the portfolio manager may be held to some index with no more than 5% tracking error. On the sell side it can be for index replication – it is cheaper to hold a smaller number of shares than to replicate the index exactly. Typically, tracking error is measured by:
$$
\text{TE} (w) = [{\text{Var}(w^T x-y)}]^{\frac{1}{2}}
$$

where $y$ is the benchmark return, $x$ is the vector of asset returns, and $w$ is the vector or weights. One way to specify the objective function is:

$$
w^* = \min_w \text{TE}(w)
$$

with

$$
\sum_{i=1}^n w_i = 1
$$

Assume that you would like to construct a tracking portfolio for the DJIA that consists of a maximum of 10 stocks (all long). You could, of course, try all 10 stock combinations of the 30 stocks in the DJIA – that would be over 30 million optimizations – so if you could run one optimization a minute, twenty-four hours a day, seven days a week, it would take more than 57 years!

#### (3) 

Just as the idea of Forward- and Backward-Stepwise Selection in linear regression, we can **select the stocks one by one and add/delete it to the set of final 10 stocks**. Here, for demonstrative purpose, we only **implement Forward Stepwise Selection**, which is a greedy algorithm, producing a nested sequence of models. In this sense it might seem sub-optimal compared to
best-subset selection. However, there are several reasons why it might be preferred: \
**Computational**: for large p (total number of factors), we cannot compute the best subset sequence, but we can always compute the forward stepwise sequence (even when p ≫ N). \
**Statistical**： a price is paid in variance for selecting the best subset of each size; forward stepwise is a more constrained search, and will have lower variance, but perhaps more bias.

```python
from sklearn.base import BaseEstimator, TransformerMixin


class TrackingErrFitter(BaseEstimator, TransformerMixin):
    def __init__(self):
        super().__init__()
        self.weights = None

    def fit(self, X, y=None):
        def tracking_error(weights, returns, benchmark):
            portf_ret = returns @ weights
            diff = portf_ret - benchmark
            te = np.sqrt(np.var(diff))
            return te

        N = X.shape[1]
        constraints = ({'type': 'eq',
                        'fun': lambda x: np.sum(x) - 1,  # weights sum = 1
                        'jac': lambda x: np.ones(len(x))})
        bounds = [(0, 1) for _ in range(N)]
        initial_guess = np.ones(N) * 1 / N
        result = minimize(tracking_error,
                          initial_guess,
                          args=(X, y),
                          method='SLSQP',
                          bounds=bounds,
                          constraints=constraints)
        self.weights = result.x

    def transform(self, X):
        return X

    def predict(self, X):
        return X @ self.weights

    def score(self, X, y=None):
        return -np.sqrt(np.var(X @ self.weights - y))
```

```python
from sklearn.feature_selection import SequentialFeatureSelector

tef = TrackingErrFitter()
sfs = SequentialFeatureSelector(tef, n_features_to_select=10, direction='forward')

# Fit the model
sfs.fit(Dec_data,Dec_ind)

# Get the selected features
selected_features = sfs.get_support(indices=True)
```

```python
selected_stocks = [i for i in np.array(Dec_data.columns)[selected_features]]
selected_stocks
```

```python
# Let's see what is the tracking error for the selected 10 stocks
tef = TrackingErrFitter()
tef.fit(Dec_data[selected_stocks],Dec_ind)
print(f'Tracking error is {-tef.score(Dec_data[selected_stocks],Dec_ind):.4f}.')
```

#### We can see that  the  tracking error from Forward Selection is much smaller than that generated from random 10 stocks.
