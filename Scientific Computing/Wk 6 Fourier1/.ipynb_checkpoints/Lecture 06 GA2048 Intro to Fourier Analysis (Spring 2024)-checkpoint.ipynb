# Overview of Fourier Analysis

Fourier Analysis is essentially the study of two fundamental objects used as a basis to represent (a large class of) functions:
1. Fourier Series studies when a function can have a trigonometric (infinite) series representation
$$ f(x) \sim \sum_{n=-\infty}^{+\infty} c_n e^{i2\pi nx} = \lim_{N\to + \infty} \sum_{n=-N}^{N} c_n e^{i2\pi nx}$$
$$ c_n = \int_0^1 f(t) e^{-i2 \pi n t} dt $$
2. Fourier Transform (continous counterpart of the Fourier Series) studies when changing a function 
$$ \F f(t) = \hat f (t) = \int_{-\infty}^{+\infty} f(x) e^{-i2\pi tx} dx = \lim_{T\to + \infty} \int_{-T}^{T} f(x) e^{-i2\pi tx} dx $$

Fourier analysis is then about:
- Finding conditions on $f$ or $c_n$ such that the (Fourier) series exanpsion is meaningful;
- Finding conditions on $f$ or $\F f$ under which we can invert the Fourier transform such as 
$ f(x) = \int_{-\infty}^{+\infty} \F f(t) e^{i 2\pi tx} dt $
- Understand the interplay between properties on $f$ (integrability, regularty, etc.) and $c_n$ (speed of type decay) and similarly, the relation between $f$ and $\F f$.

### Applications (seen in finance and not limited to)
- Function Approximation / Smoothing
- Times series Analysis/Denoising
- Dimensionality Reduction
- Factor Analysis
- Statistical Analysis
- Numerical Analysis: PDE solving, Derivatives Pricing Techniques

# Fourier Series

## Introduction
Let's consider an integrable function $f$ assuming a trigonometric (i.e. Fourier) series representation holds, namely one has
$$ f(x) = \sum_{n=-\infty}^{+\infty} c_n e^{i2\pi nx} $$

where $e^{iz} = \cos(z) + i \sin(z)$. Then $f$ is a 1-periodic function: $f(x+1) = f(x+k)$ for all $k \in \mathbb{Z}$. 


A first consideration would be to understand how to compute the coefficient $c_n$? In "Théorie Analytique de la Chaleur" Fourier's method (for some special functions like $f = 1_{[0,1]}(x)$ on $(-1, 1)$) was to solve the linear system of the first N variables, assuming $f$ is differientiable $N$ times at $x=0$, then letting $N \to +\infty$. Really not easy, works essentially on special examples, but Fourier soon found there was an expeditive method, although he himself noted in his work that knowing how to compute these coefficients would in no way indicates if the corresponding trigonometric series converges whatsoever.

Recall that:
$$ \int_0^1 e^{i2 \pi n x} dx = \begin{cases}
    1 & \text{if} \; n= 0\\
      0 & \text{if} \; n \neq 0\\
\end{cases}   
$$

Let's assume that $\sum_{n \in \Z} |c_n| < +\infty$ so that the sum $\sum_{n=-\infty}^{+\infty} c_n e^{i2\pi nx}$ converge uniformly on $\R$. Then we can interchange sum and integral, and we get:
$$ \int_0^1 f(x) e^{-i2\pi m x} dx = \sum_{n=-\infty}^{+\infty} c_n \int_0^1 e^{i2\pi (n-m)x} dx = \sum_{n=-\infty}^{+\infty} c_n \delta_m(n) = c_m $$

That indicates that if such a representation holds, we should have:
$$ c_n(f) = \hat f(n) = \int_0^1 f(x) e^{-in 2 \pi x} dx $$

As such, for a given integrable function $f$ we can define its **Fourier coefficients** as 
$$ c_n(f) := \hat f(n) = \int_0^1 f(x) e^{-i2 \pi n x} dx $$

We note only $c_n$ when it's clear from the context. The notation $\hat f(n)$ is suggestive of how it looks like the Fourier transform of $f$ at integer values. Then one can define the partial **Fourier Series** of $f$:
$$ S_N^f(x) = \sum_{n=-N}^N c_n(f) e^{i 2 \pi n x} $$

It is not clear *a priori* if $S_Nf(x)$ converges or not. In fact, even if $f$ is continuous, its Foufier Series might diverge as shown by [Fejer](https://www.mathcounterexamples.net/continuous-function-with-divergent-fourier-series/). Still, one **formally write**:
$$ f(x) \sim \sum_{n=-\infty}^{+\infty} c_n e^{i2\pi nx} $$

to indicate that the correspondance between a function $f$ and its Fourier Series.

The fact that if a function has a trigonometric series representation, then its coefficients must have a certain form also suggests some uniquesness in this representation. In particular, we collect below a few general results:



## Remark on the Frequency

What if we consider a function with a period different from 1? If $f$ is a $T$-periodic function, then $g(x) = f(Tx)$ is a 1-periodic function and we repeat the same analysis an $g$ and revert it as $f(x) = g(x/T)$.

If $f$ is a T-periodic function:
$$ c_n(f,T) = \frac{1}{T} \int_0^T f(x) e^{-i \frac{2 \pi}{T}n x} dx = \frac{1}{T} \int_{-T/2}^{T/2} f(x) e^{-i \frac{2 \pi}{T}n x} dx $$
$$ f(x) \sim \sum_{n \in Z} c_n(f,T) e^{i \frac{2 \pi}{T}n x} $$

More generally, if $f$ is $p_1$-periodic, then $g(x) = f(p_1/p_2 x + b)$ is a $p_2$-periodic function for any $b$.

Also, if $f$ is a function define on an interval $I$ of length $L$, then the **periodization** of $f$ with period $L$ is the function $f_L(x) = \sum_{n \in Z} f(x + nL)$.

## Cosine and Sine representation of Fourier Series
From $e^{i \theta} = \cos(\theta) + i \sin(\theta)$ one easily verifies that
$$ 2\int_0^1 f(x) \cos(2\pi n x) dx = c_n + c_{-n} = a_n $$
$$ 2\int_0^1 f(x) \sin(2\pi n x) dx = i(c_n - c_{-n}) = b_n $$

and 
\begin{split}
\sum_{n=-\infty}^{+\infty} c_n e^{i2\pi nx}&= c_0 + \sum_{n=1}^{+\infty} c_n e^{i 2 \pi nx} + \sum_{n=1}^{+\infty} c_{-n} e^{-i 2 \pi nx} \\
& = \frac{a_0}{2} + \sum_{n=1}^{+\infty} a_n \cos(2\pi n x) + \sum_{n=1}^{+\infty} b_n \sin(2\pi n x) 
\end{split}

From the definition of $c_n$, one readily sees that $\overline{c_n(f)} = c_{-n}(\bar f)$, so that  if $f$ is a real-valued function, then $\overline{c_n} = c_{-n}$ and as such
$$ c_n + c_{-n} = 2 Re(c_n) = a_n $$
$$ i(c_n - c_{-n}) = -2 Im(c_n) = b_n $$

so that
$$ c_n = \half( a_n - ib_n) $$

Note that for any real-value function $f$, if one defines and even function $f_{even}(x) = \frac{f(x)+f(-x)}{2}$ and an off function $f_{odd}(x) = \frac{f(x) - f(-x)}{2}$, such that $f(x) = f_{even}(x) + f_{odd}(x)$. This decomposition is unique, and therefore, if the above trigonometric repesentation holds, then \
$$f_{even}(x) = \frac{a_0}{2} + \sum_{n=1}^{+\infty} a_n \cos(2\pi n x) \;\;\text{and}\;\; f_{odd}(x) = \sum_{n=1}^{+\infty} b_n \sin(2\pi n x) $$

This is useful as when we know whether $f$ is odd or even, then $a_n \equiv 0$ or $b_n \equiv 0$ respectively. This representation can also be convenient when $f$ is real-valued.

## Theory of Convergence of Fourier Series

There are essentially two view poitns to start from:
1. From the properties of a sequence $\lambda_n$, when can we infer on $\sum_{n\in \Z} \lambda_n e^{inx}$. 
2. From the properties of a function $f$, what can we infer on $S_N^f$.

There is a duality between the regularity of $f$ and the speed at which the coeeficients of its Fourier Series decreases. In particular, by doing successive integration by parts, one can see that as $n \to +\infty$
$$ c_n(f) = \frac{1}{(in)^k} \int_0^1 f^{(k)}(x) e^{-i2 \pi nx} dx = O\left(\frac{1}{n^k} \right) \stackrel{Riemann-Lebesgue}{=} o\left(\frac{1}{n^k} \right)$$

If $f$ is $\a$-Holder regular (i.e. $|f(u) - f(v)| \leq C |u-v|^{\alpha}$), then (Theorem of Berstein), then $c_n(f) = O(1/n^\a)$



### Main results

#### Properties from a given sequence:
* If $\sum_{n \in Z} |\lambda_n| < +\infty$, then $\sum_{n=-N}^N \lambda_n e^{i 2 \pi nx}$ converges **absolutely** (and therefore **uniformly**) to a 1-perodic function. In which case, $\lambda_n = c_n(f)$.
* If $f$ is intregrable over $[0,1]$ and $c_n(f) = 0$ for all $n \in Z$. Then $f(x) = 0$ whenever $f$ is continuous at $x$.
* Theorem of Lebesgue-Cantor: If $\forall x \in \R \;$  $\lim_{n\to +\infty} c_n e^{-inx}+c_n e^{-inx} = 0$ then $ \lim_{|n| \to +\infty}c_n = 0$.
* Theorem of Cantor: If $\forall x \in \R \;$  $\lim_{N\to + \infty} \sum_{|n| \leq N} c_n e^{inx} = 0$ then $c_n = 0$.

#### Properties from a given function $f$:
* If $f$ is continuous and $c_n(f) = 0$ for all $n$ then $f \equiv 0$.
* If $f$ is continuous and $\sum_n |c_n| < +\infty$ then $S_N^f(x) \to f(x)$ **uniformly** in x.
* Theorem of Rieman-Lebesgue: if $f \in L_1$ then $c_n(f) \to 0$ as $|n| \to +\infty$.
* If $f$ is integrable and differientiable at $x$ then $S_N^f(x) \to f(x)$ pointwise.
* If $f$ is contiuous and piecewise $C^1$, then $ S_N^f(x) \to f(x)$ absolutely.
* If $f$ is $\a$-Holder with $\a > \half$, then $S_N^f \to f$ absolutely.

We note $f(x^+) = \lim_{\eps \downarrow 0} f(x+\eps)$ and $f(x^-) = \lim_{\eps \downarrow 0} f(x-\eps)$.

We have for $f$ piecewise-$C_1([0,1])$:
* $S_N^f(x) \to \frac{f(x^+) + f(x^-)}{2}$
* Defining $g(x) = \frac{f'(x^+) + f'(x^-)}{2}$, then $c_n(g) = i n c_n(f)$.

### Some Important (Counter) Examples

* We have already mention the exemple from [Fejer](https://www.mathcounterexamples.net/continuous-function-with-divergent-fourier-series/) of a $2\pi$-periodic continous function, $f(x) = \sum_{n=1}^{+\infty} \frac{\sin\left(\frac{x}{2}(2^{n^3}+1)\right)}{n^2}$, whose Fourier series diverges at $x=0$.

* For $0 < \a < 1$, the function $\sum_{n \geq 1} \frac{\sin nx}{n^\a}$ converges for every $x$ but is not the Fourier series of an integrable function.

* The function $\sum_{n \geq 2} \frac{\sin nx}{\log(n)}$ converges for every $x$ but is not the Fourier series of an integrable function.

* The function $f(x) = \sum_{n=1}^{+\infty} \frac{\sin nx}{n^2}$ is continuous with its Fourier coefficients $b_n(f) = 1/n^2$ and therefore satisfy $\sum_{n \geq 1} |n b_n(f)|^2 < +\infty$ but $f$ is not differentiable at $x=0$. 

## Kernels

The $N^{th}$ **Dirichlet kernel** is defined as:
$$ D_N(x) = \sum_{k=-N}^N e^{i k x} = \frac{\sin\left( \left(N + \half \right)x\right)}{\sin(x/2)} $$

The **Poisson kernel** is defined for $\theta \in [-\pi, \pi]$ and $0\leq r < 1$ as:
$$ P_r(\theta) = \sum_{n\in \Z} r^{|n|} e^{in \theta} = \frac{1-r^2}{1-2\cos(\theta) + r^2} $$

The $N^{th}$ **Fejer kernel** is defined as:
$$ F_N(x) = \frac{1}{N} \sum_{n=0}^{N-1} D_i(x) = \frac{1}{N} \frac{\sin^2(Nx/2)}{\sin^2(x/2)} $$


## Convolution

The convolution of two $2\pi$-periodic functions is defined as:
$$ (f*g)(x) = \frac{1}{2\pi} \int_{-\pi}^\pi f(t) g(x-t) dx $$

More generally, the  convolution of two $T$-periodic functions is defined as:
$$ (f*g)(x) = \frac{1}{T} \int_{-T/2}^{T/2} f(t) g(x-t) dx $$

Noting $\tilde D_N(x) = D_N(2\pi x)$, this gives when considering $1$-periodic functions:
\begin{split}
 S_N f(x) & = \sum_{n=-N}^N c_n(f) e^{i 2 \pi n x}\\
 & = \sum_{n=-N}^N \left(\int_0^1 f(t) e^{-in 2 \pi t} dt \right) e^{i 2 \pi n x} \\
& = \int_0^1 f(t) \sum_{n=-N}^N e^{i 2 \pi n (x-t)} dt \\
& = (f* \tilde D_N)(x)
\end{split}

Key properties of convolutions are:
$$ (f*g)(x) = (g*f)(x)  $$
$$ \widehat{(f*g)}(n) = \hat f(n) \hat g(n) $$

## Fourier-Fejer Summation

Instead of looking at the convergence $S_N^f(x) \to f(x)$ pointwise, Fejer in 1900 showed that instead the convergence in the Cesaro summability sense yields very interesting results. Let's consider
$$ \s_N^f (x) = \frac{1}{N} \sum_{n=0}^{N-1} S_n^f(x)$$

Then if $f$ is integrable, then $\s_N^f(x) \to f(x)$ at every point where $f$ is continuous. But more importantly, if $f$ is **continuous**, then $\s_N^f(x) \to f(x)$ **uniformly**!

It is useful to see that:
\begin{split}
\s_N^f(x) &= \frac{1}{N} \sum_{k=0}^{N-1} \sum_{n=-k}^k c_n e^{i2\pi nx} \\
&= \sum_{n=-\infty}^{+\infty} \frac{1}{N} \sum_{k=0}^{N-1} \left( 1_{\{ |n| \leq k \}}\right) c_n e^{i 2\pi nx}
\end{split}

but since $ \sum_{k=0}^{N-1} 1_{\{ |n| \leq k \}} = (N-|n|)1_{\{|n| \leq N\}}$, we end up with
$$ \s_N^f(x) = \sum_{n=-N}^{N} \left( 1 - \frac{|n|}{N}\right) c_n e^{i 2\pi nx} $$

This expression is (almost) like the original Fourier series $S_N^f$ but with the weight $\w_n^{(N)}=\left( 1 - \frac{|n|}{N}\right)$ which will add smoothing.

As a remark, this summation is linked to the (1-periodized) Fejer kernel as
$$ \s_N^f(x) = (f * F_N)(x) $$

## $L^2$ Theory of Fourier Series

There is an extensive body of textbooks and scientific literature on Hilbert space and the fact that the family of function $(e^{inx})_{n \in \Z}$ form a complete basis of $L^2([0,1])$. In the context of Fourier Series, the main results are the $L^2$ convergence theorem and Parseval identities.

### $L^2$ convergence

If $f \in L^2([0,1])$ then as $N \to +\infty$
$$ \int_0^1 \left|f(x) - S_N^f(x) \right|^2 \to 0  $$

In particular, since $[0,1]$ is of finite length, the result holds in $L^1$ and there is a subsequent $\phi(N)$ such that $S_{\phi(N)}^f(x) \to f(x)$ pointwise.



### Parseval indentiy

$$ \int_0^1 |f(x)|^2 dx = \sum_{n\in Z} |c_n|^2 = \frac{|a_0|}{4} + \half \sum_{n\geq 1} \left(|a_n|^2 + |b_n|^2 \right)$$

More generally, for two $L^2$ function $f$ and $g$, then
$$ \int_0^1 f(x) \bar{g}(x) dx  = \sum_{n \in \Z} c_n(f) \bar{c}_n(g) $$


### Least-Square Approximation

For any linear combination with complex numbers $\lambda_n$, one has
$$ || f - \sum_{n=-N}^{N} c_n(f) e^{in2\pi x} ||_2 = || f - S_N^{f} ||_2 \leq || f - \sum_{n=-N}^{N} \lambda_n e^{in2\pi x} ||_2 $$

That is the Fourier coefficients achieve the best least-square approximation of 

In particular, from $M = 2N+1$ observation of $f(x_i)$, a naive least-square inversion (solving and inverting the normal equations) is an $O(M^3)$ algorithm. A naive computation of the Fourier-coefficients is an $O(M^2)$ complexity. The Fast Fourier Transform, is an $O(M \log(M))$ algorithm.

# Discrete  Fourier Transform (DFT) and Fast Fourier Transform (FFT)

In practice, all continuous problems are discretized. If we were to try to numerically approximate the Fourier coefficients of a function $f$, a natural way to proceed is to write:
$$ c_k = \int_0^1 f(x) e^{-i 2\pi k x} dx \approx \frac{1}{N} \sum_{n=0}^{N-1} f\left(\frac{n}{N}\right) e^{-i 2\pi n k/N} $$

Using more generic notation, it is therefore natural to consider a finite sequence of observed values, let's note it $x(n) $, and find a similar expression for it's Fourier coefficient. Let's consider

$$ \hat x(k) = \frac{1}{N} \sum_{n=0}^{N-1} x(n) e^{-i2\pi nk/N} $$

Note that $e^{i2\pi n/N}$ is the $n^{th}$ root of unity. Noting $w_N = e^{i2\pi/N}$, then the solutions of $z^N=1$ are $1, w_N, w_N^2,...,w_N^{N-1}$ and taking higher (or negative) powers of $w_N$ amounts to keep going around the unit circle. For instance, $w_N^{-k} = w_N^{N-k}$ for $0 \leq k \leq N$. 

In the same way, $\hat x(-k) = \hat x(N-k)$ and if $N$ is odd for instance, then the Discrete Fourier transform amounts at computing $\hat x\left(-\frac{N-1}{2}\right),..., \hat x(0), ..., \hat x\left(\frac{N-1}{2}\right)$. 

This is why, just like pieces of functions on a finite interval are extended as a function on $\R$ by periodicity, we can think of a sequence $x(0),...,x(N-1)$ as a sequence in $\Z/N\Z$, similarly to what happens on the roots of unity.

Just like for Fourier series, simple algebra gives
$$  \sum_{k=0}^{N-1} e^{i 2\pi k n/N} = \begin{cases}
    N & \text{if} \; n= 0\\
      0 & \text{if} \; n \neq 0\\
\end{cases}  $$


which yields the main **inversion** formula:
$$ x(n) = \sum_{k=0}^{N-1} \hat x(k) e^{i2\pi nk/N}$$

and Plancherel indentity:
$$ \sum_{n=0}^{N-1} |x(n)|^2 = \sum_{k=0}^{N-1} |\hat x(k)|^2  $$

<div class="alert alert-block alert-info">

- Somtimes the DFT definitions are different and the factor $1/N$ is not in the definition of $x(k)$ and is therefore needed in the inversion-formula, or the factor $1/\sqrt{N}$ is used in both the definition of $\hat x{k}$ and the inversion formula.
    
- By periodicity (or simply changing the indices, if $N$ is even, the definition can be expressed as:
$$ \hat x(k) = \sum_{n=-N/2-1}^{N/2} x(n) e^{-i2\pi nk/(N+1)} $$

## Fast Fourier Transform

The Fast Fourier Transform ([FFT](https://en.wikipedia.org/wiki/Fast_Fourier_transform)) is an efficient algorithm to compute the DFT. The key element comes from the periodicity of the unit roots around the circle. 

Noting $w_N = e^{-i2\pi/N}$, and storing once and fall all $k=1,...,N$ the various power $w_n^k$ at compile time once and for all in the heap, then one needs to compute the value $\hat x(k)$, also for $k=1,...,N$. The calculations of each $\hat x(k)$ requires muliplying and adding $N$ terms. Taking this operation as the performance unit, a brute force approach for the DFT calculation is in $O(N^2)$.

The FFT is an algorithm that enable one to reduce the calculation complexity to $O(N \log(N))$ and has been considered as the most important breakthrough in numerical computing of the 20$^{th}$ century. The famous [FFTAlgo](https://en.wikipedia.org/wiki/Cooley%E2%80%93Tukey_FFT_algorithm) can be essentially summarized from the case $N=2^n$, because if $N$ is even then:
\begin{split}
\hat x(k) &= \frac{1}{N}\sum_{n=0}^{N/2} x(2n) e^{-i \frac{2\pi}{N} 2n k} + e^{-i \frac{2\pi}{N}k} \frac{1}{N}\sum_{n=0}^{N/2} x(2n+1) e^{-i \frac{2\pi}{N} 2n k} \\
&= \frac{1}{N}\sum_{n=0}^{N/2} x(2n) e^{-i \frac{2\pi}{N/2} nk} + e^{-i \frac{2\pi}{N}k} \frac{1}{N}\sum_{n=0}^{N/2} x(2n+1) e^{-i \frac{2\pi}{N/2} n k} \\
& = \hat e(k) + e^{-i \frac{2\pi}{N}k} \hat o(k)
\end{split}

where $\hat e (k)$ and $\hat o(k)$ are respectively a DFT on even and off indices of the original vector of points. The next key observation is that $\hat e(k+N/2) = \hat e(k)$ and $\hat o(k+N/2) = \hat o(k)$ because the $N/2$ cancels out and $e^{-i \frac{2\pi}{N}(k+N/2)} = e^{-i \frac{2\pi}{N}k}e^{-i\pi} = - e^{-i \frac{2\pi}{N}k}$ which yields
$$ \hat x(N/2 + k) = \hat e(k) - e^{-i \frac{2\pi}{N}k} \hat o(k)$$

As such, once $\hat e(k)$ and $\hat o(k)$ are computed  with half of the full calculations for $k=0,...,N/2-1$, then all the $\hat x(k)$ can be recovered.

By induction over $n$ we get the $\log(N)$ complexity for each calculation of $\hat x(k)$ hence a total complexity of order $O(N \log(N))$.

### FFT  and Convolution

The calculation of vectors $(v(k))_{0\leq k \leq N-1}$ of the form
$$ v(k) = \sum_{n=0}^{N-1} a_n b_{k-n} $$

can be performed efficiently using the FFT algorithm.

### Remark on real signals and coefficients symmetry

If $f$ is a real function, although Fourier analysis usually involves complex numbers, the cosine and sine formulation of the Fourier series shows how everything can be dealt with real numbers.

When dealing with DFT, what we have is
$$ x(n) = \overline{x(n)} \Rightarrow \overline{\hat x(k)} = \hat x(-k) = \hat x(N-k) $$

# Fourier Transform

## Intuition Behind Definition and Fourier Inverse Formula
If $f$ is not periodic, one can consider the periodization of $f$ restricted to an interval $[-L/2, L/2]$ and extended by periodization, namely $f_L(x) = \sum_{n\in\Z} f(x+nL) 1_{\left[-\frac{L}{2}, \frac{L}{2}\right] }(x+nL)$. Trivially, one has $f_L \to f$ as $L \to +\infty$.

Then one can write the Fourier series expansion of $f_L$, which would yield:
$$ f_L(x) = \frac{1}{L} \sum_{n\in\Z} \hat f_L\left( \frac{n}{L} \right) e^{i 2 \pi\frac{n}{L} x} $$

where 
$$\hat f_L(z) = \int_{-L/2}^{L/2} f_L(x) e^{-2i\pi x z}dx = \int_\R f(x)1_{\left[-\frac{L}{2}, \frac{J}{2}\right]}(x) e^{-2i\pi x z} dx $$

so that under if $f \in L^1(\R)$, then $\hat f_L(z) \to \hat f (z)$ as $L \to +\infty$.

Remarking that (under appropriate assumptions) 
$$ \int_\R g(t) dt = \lim_{ \delta \downarrow 0} \delta \sum_\Z g(n \delta) $$

one can see, by setting $\delta = 1/L$, that the limit of $f_L$ as $L\to +\infty$, yields:
$$ f(x) = \int_\R \hat f(t) e^{i2\pi xt} dt $$


## Definition of Fourier Transform
The Fourier transform of an integrable function is defined as
$$ \F f(t) = \hat f(t) = \int_{\R} f(x) e^{i2\pi x t} dx$$

The most important result is that the Fourier transform contains all the information about $f$ so that if $f$ is continuous or in $L^1(\R)$ and $\F f \in L_1(\R)$, then
$$ f(x) = \int_{\R} \F f(t) e^{-i2\pi x t} dt $$

So that if one defines the inverse Fourier transform as
$$ \F^{-1} f(t) = \int_{\R} f(x) e^{-i2\pi x t} dx $$

With this definition, one sees that $\F \F^{-1} = \F^{-1}\F = Id$ so that definine the Fourier transform as $\hat f(t) = \int_{\R} f(x) e^{-i2\pi x t} dx$ is equivalent, one just needs to be careful of the sign.

### Remarks on equivalent definitions
Depending on the context (i.e. in Probability theory) and/or preferences, the Fourier transform of $f$ is sometimes defined as
$$ \tilde \F f(t) = \int_{\R} f(x) e^{i x t} dx = \F f \left(\frac{t}{2 \pi} \right) $$

From the first Fourier transform formula presented, one sees from a simple change of variable that
$$ f(x) = \int_{\R} \F f(t) e^{-i2\pi x t} dt = \frac{1}{2\pi} \int_{\R} \F f \left(\frac{t}{-2 \pi} \right) e^{i x t}  dt = \frac{1}{2\pi} \int_{\R} \tilde \F f (t) e^{-i x t}  dt $$

so that in fact, one simply has
$$ \tilde \F^{-1}f(x) = \frac{1}{2\pi} \int_{\R} f (t) e^{-i x t} dt$$

Depending on the context, Fourier transform can either mean $\F f$ or $\tilde \F f$ and the Inverse Fourier transform formula would simply correspond accordingly.

Similarly, defnining the Fourier transform as
$$ \bar \F f(t) = \int_{\R} f(x) e^{-i x t} dx = \F f \left(\frac{-t}{2 \pi} \right) $$
yields
$$ \bar \F^{-1} f(t) = \frac{1}{2\pi} \int_{\R} f(x) e^{i x t} dx = \F f \left(\frac{t}{2 \pi} \right) $$

Finally, sometimes the scaling factor $\frac{1}{\sqrt{2 \pi}}$ is found in the definition of $\tilde \F$ or $\bar F$ so as the make the inverse transform formula more symmetrical with only a sign flip.

## Basic Results on Fourier Transform

Main classic results on the Fourier transform are:
- If $f \in L^1(\R)$, then its Foutier transform $\hat f$ is bounded and uniformly continous
- $f \in L^1(\R) \Rightarrow \lim_{|t| \to +\infty} |\hat f(t)| = 0$
- if $x^k f(x) \in L^1$ then $\F\left((-i2\pi)^k f(t)\right) = \hat{f^{(k)}} $
- If $f \in \mathcal C^k$ then $\F f^{(k)} = (i2\pi)^k \hat f$ and $\hat f(t) = o(t^{-k})$
- For $f,g \in L^1(\R)$ then $\sup_t |\hat f(t) - \hat g(t)| \leq \int_\R |f(x) - g(x)| dx$ 
- If inversion formula holds then $\sup_x |f(x) - g(x)| \leq \int_\R |\hat f(t) - \hat g(t)| dt$

## Examples of Fourier Transforms

- $f(x) = 1_{\left[-\frac{T}{2}, \frac{T}{2}\right]}(x) \to \hat f(t) = \frac{\sin(\pi t T)}{\pi T}$
- $f(x) = e^{-ax}1_{\R^+}(x) \to \hat f(t) = \frac{1}{a + i2\pi t}$

Another important example, linked to the sampling formula is:
$$ f_n(x) = \frac{\sin(\pi(x-n))}{\pi(x-n)} \Leftrightarrow \hat f_n(t) = 1_{\{|t| \leq \half \}} e^{-i2 \pi n t} $$

## Numerical Computation

One can think of 2 ways to proceed:
1. From the Inverse Fourier formula:
$$ f(x) = \int_{\R} \hat f(t) e^{2\pi tx} dt $$

Then for a small discretization step $\delta$, one can use
$$ f(x) \sim \delta \sum_{n\in \Z} \hat f(n\delta) e^{2\pi n \delta x} $$

Then the truncation approximation becomes
$$ f(x) \sim \delta \sum_{k=-N}^N \hat f(k\delta) e^{2\pi k \delta x} $$

2. Still from the inverse formula we can write
$$ f(x) = \int_{\R} \hat f(t) e^{2\pi tx} dt \sim \int_{-T}^T \hat f(t) e^{2\pi tx} dt $$

and choosing $N$ large enough, one gets
$$ f(x) \sim  \frac{2T}{N} \sum_{n=-N}^N \hat f\left(-T + k\frac{2T}{N}\right) e^{2\pi k\frac{2T}{N}x}  $$

Choosing $\delta = \frac{2T}{N}$ yields equivalence between the two approaches.

In order to apply the FFT algorithm, we need to choose values $x_n$ at which the function $f$ will be approximated. 

### Some Tricks
Sometimes it is useful, especially when $\hat f$ is symmetrical, to write
$$ f(x) = \int_0^{+\infty} \left(\hat f(t) e^{i2\pi tx} +\hat f(-t) e^{-i2\pi tx} \right) dt $$

If $\hat f$ is symmetrical, then $f$ is real-valued as:
$$ f(x) = 2 \int_0^{+\infty} \hat f(t) \cos(2\pi tx) dt $$

If $f$ is **real-valued**, then $\overline{ f(x)} = f(x)$ for all $x$ which yields $\overline{\hat f(t)} = \hat f(-t)$ which gives
$$ f(x) = \int_0^{+\infty} \left(\hat f(t) e^{i2\pi tx} +\overline{\hat f(-t) e^{-i2\pi tx}} \right) dt = 2  \int_0^{+\infty}Re\left( \hat f(t) e^{i2\pi tx} dt \right) = 2 Re\left( \int_0^{+\infty} \hat f(t) e^{i2\pi tx} dt \right) $$

### Another thing to mention
Let's consider the integral that needs to be calculated has the form
$$ f(x) = \int_0^{+\infty} g(t) e^{-itx}dt \sim \int_0^{T} g(t) e^{-itx}dt $$

Writing $\Delta T_N = \frac{T}{N}$, a discretization of the integral at the points $t_k = k\Delta T_N$ gives
$$ f(x) \sim \sum_{k=0}^{N-1} g(k\Delta T_N) e^{-i k \Delta T_N x}\Delta T_N $$

writing $x_n = -x_{low} + n\Delta_x$, with $\Delta_x = (x_{high}-x_{low})/N$ and $n \in \{0,...,N-1 \}$ leads to
$$ f(x_n) \sim \Delta T_N\sum_{k=0}^{N-1} g(k\Delta T_N) e^{i k \Delta T_N x_{low}} e^{-i k n\Delta T_N \Delta_x} $$

and then the discretization steps are chosen so that $\Delta T_N \Delta_x = 2 \pi / N$ so that the FFT can now be applied with $x(k) = g(k\Delta T_N) e^{i k \Delta T_N x_{low}}$ as then
$$ f(x_n) \sim T \frac{1}{N} \sum_{k=0}^{N-1} x(k) e^{-i 2 \pi k \frac{n}{N}} $$

<div class="class-alert alert-block alert-info">
<b>Remark:</b> Depending on the calculation needed, the integration boundary can be different. Each situation might require care on how the truncation is put in place.
</div>

## Numerical Example of Fourier Transform Computation
We want to verify the identity
$$ f(x) = \int_{\R} e^{it x} e^{-\half t^2} dt = \sqrt{2 \pi} e^{-\half x^2} $$

For $\hat f(t) = e^{-\half t^2} $, $t_k = -T + k \Delta t$, $\Delta t = \frac{2T}{N}$ and $x_n = -x_{low} + n \Delta_x$ with $\Delta_x = 2 x_{low}/N$, we will use the following approximation
$$ f(x_n) \approx \int_{-T}^T e^{it x} \hat f(t) dt \approx \Delta t \sum_{k=0}^{N-1} e^{i (x_{low} + n \Delta_x)(-T + k \Delta t)} \hat f(t_k) = \Delta t e^{- i nT \Delta_x} \sum_{k=0}^{N-1} e^{i x_{low}t_k } e^{i \Delta_t \Delta_x n k} \hat f(t_k) $$

Imposing $\Delta_x \Delta t = 2 \pi /N$ yields $x_{low} = N \frac{\pi}{2T}$ and 
$$ f(x_n) \approx 2T e^{- i nT \Delta_x} \frac{1}{N} \sum_{k=0}^{N-1} \hat x(k)  e^{i 2 \pi n k/N}$$

$$\hat x(k) = e^{i x_{low}t_k } \hat f(t_k)$$

```python
# Gaussian Fourier Transform
N = 100
T = 5.
Delta_T = 2*T/N
x_min = N*np.pi/(2*T)
Delta_x = 2*x_min/N

x_n = [-x_min + n*Delta_x for n in range(N)]
t_k = [-T + k*Delta_T for k in range(N)]

f_hat = lambda x: np.exp(-0.5*x*x)
f_hat_value = [f_hat(t) for t in t_k ]

sqrt_2_pi = np.sqrt(2*np.pi)

f_target = lambda x: sqrt_2_pi * np.exp(-0.5*x*x)
f_target_value = [f_target(x) for x in x_n]

figure(figsize=[14, 5])
subplot(1,2,1)
plot(t_k, f_hat_value)
title("$\hat f (x) = e^{- x^2/2} $")

subplot(1,2,2)
plot(x_n[int(0.3*N):int(0.7*N)], f_target_value[int(0.3*N):int(0.7*N)])
title("$ f (x) = \sqrt{2\pi} e^{- x^2/2} $")
```

```python
# using FFT
x_hat = [np.exp(complex_i * x_min*t)*f_hat(t) for t in t_k]
f_x_fft = ifft(x_hat)

f_from_fft = np.array([2*T*np.exp(-complex_i*n*np.pi) for n in range(N)]) * np.array(f_x_fft)

print("verifying the imaginary part is numerically zero")
print(np.sum( np.abs([f.imag for f in f_from_fft])) )

f_from_fft_real = [f.real for f in f_from_fft]

figure(figsize=[14, 5])

subplot(1,2,1)
plot(x_n[30:70], f_from_fft_real[30:70])
title("$f(x) = \int \hat f(t) e^{itx} dt$ from DFT approximation")

subplot(1,2,2)
plot(x_n[30:70], np.array(f_from_fft_real[30:70]) - np.array(f_target_value[30:70]))
title("Error: $f(x) - f_{from\;fft}(x)$")
```

The recovery appears to be quite precise

### Using artificial deterministic noise

Let's reconsider the earlier FFT example, but this time we add some noise. In this simple example, noise is deterministic and is "pretended" to be a harmonic with a very high frequency.


```python
# define a nice 2pi-periodic signal
f = lambda x: np.cos(3 * x) / 3 + np.cos(13 * x) / 2 + np.sin(5 * x) * 2 + np.sin(17 * x) * 1.5

# number of observation
nbPoints = 1000
X = np.linspace(0, 2*pi, nbPoints)

# get the values of the signal on [0, 2pi]
f_x = np.array([f(x) for x in X])

# define some noise
g = lambda x: np.sin(100*x)
noise = np.array([0.5*g(x) for x in X])

# get a noisy version of the original signal
f_noisy = f_x + noise

figure(figsize=[16, 5])
subplot(1, 3, 1)
title("original nice signal")
plot(X, f_x)

subplot(1, 3, 2)
title("sine function with high frequency")
plot(X, noise)

subplot(1, 3, 3)
title("(deterministically) noisy signal")
plot(X, f_noisy)
```

```python
# frequency from noisy signal
Ff_x_noisy = fft(f_noisy)

xf = fftfreq(nbPoints, 1/nbPoints)

figure(figsize=[7, 5])
plot(xf, np.abs(Ff_x_noisy))
```

One can see the extra frequency at level 100

```python
def builtRealSignal(x, freq, a, b):
    res = 0
    for i, f in enumerate(freq):        
        if i < int(len(freq)/2):
            res += a[i]*np.cos(f*x)
            res += b[i]*np.sin(f*x)
 
    return res
```

```python
# Recovering the original signal from noisy signal 
noise_freq_threshold = 50  # frequency from which a signal is considered noise - ad hoc for illustration purposes

# Manual extraction of frequencies used this time to remove noise
freq_no_noise = []
for i, r in enumerate(Ff_x_noisy):
    if np.abs(r)>50 and i < noise_freq_threshold:
        freq_no_noise.append(i)

# getting real coefficients
a_no_noise = [((Ff_x_noisy[f]+Ff_x_noisy[-f])/nbPoints).real for f in freq_no_noise[:int(len(freq_no_noise)/2)] if np.abs(((Ff_x_noisy[f]+Ff_x_noisy[-f])/nbPoints).imag) < 0.1]
b_no_noise = [(complex_i*(Ff_x_noisy[f]-Ff_x_noisy[-f])/nbPoints).real for f in freq_no_noise[:int(len(freq_no_noise)/2)] if np.abs((complex_i*(Ff_x_noisy[f]-Ff_x_noisy[-f])/nbPoints).imag) < 0.2]

rebuilt_signal_no_noise = [builtRealSignal(x, freq_no_noise, a_no_noise, b_no_noise) for x in X]
#print(rebuilt_signal)
figure(figsize=[15,4])
subplot(1,2,1)
plot(X, rebuilt_signal_no_noise)
plot(X, f_x)
subplot(1,2,2)
plot(X, f_x-rebuilt_signal_no_noise)
```

# Poisson Summation Formula

If $f \in L^1$, then for $T>0$ the function $g(t) = \sum_{n \in \Z} f(t + nT)$ is defined by absolute convergence as
\begin{split}
\int_0^T |g(t)| dt &\leq \int_0^T \sum_{n\in\Z} |f(t+nT)| dt = \sum_{n\in\Z} \int_0^T |f(t+nT)| dt  \\
& = \sum_{n\in\Z} \int_{nT}^{(n+1)T} |f(t)| dt = \int_\R |f(t)| dt < +\infty
\end{split}

Besides, $g$ is $T$-periodic and $c_n(g) = \frac{1}{T}\hat f \left(\frac{n}{T}\right)$ by direct computation:
\begin{split}
c_n(g) & = \frac{1}{T} \int_0^T \sum_{n\in\Z} f(t+nT) e^{-i2\pi \frac{n}{T}t} dt \\
&= \frac{1}{T} \int_0^T \sum_{n\in\Z} f(t+nT) e^{-i2\pi \frac{n}{T}(t+nT)} dt \\
& = \frac{1}{T} \sum_{n\in\Z} \int_{nT}^{(n+1)T}  f(t) e^{-i2\pi \frac{n}{T}t} dt \\
& = \frac{1}{T} \hat f \left(\frac{n}{T}\right)
\end{split}

As such, if $\sum_{n \in \Z} \left| \hat f \left(\frac{n}{T}\right) \right| < +\infty$, then 
$$ \boxed{\sum_{n \in \Z} f(t + nT) = \frac{1}{T} \sum_{n \in \Z} \hat f \left(\frac{n}{T}\right) e^{i 2\pi \frac{n}{T}  t}}$$

From $\hat {\hat f}(x) = f(-x)$, the formula can also be read as (note the minus sign in the r.h.s.)
$$ \boxed{\sum_{n \in \Z} \hat f(t + nT) = \frac{1}{T} \sum_{n \in \Z} f \left(\frac{n}{T}\right) e^{-i 2\pi\frac{n}{T}  t}}$$

Writing $\delta = 1/T$, the formula reads
$$ \sum_{n \in \Z} f\left(t + \frac{n}{\delta}\right) = \delta \sum_{n \in \Z} \hat f \left(n\delta\right) e^{i2\pi n \delta t}  $$

# Sampling Theorem (Shannon-Nyquist)

Let's consider a function $f$ whose Fourier transform $\hat f$ is null outside of an interval $[-F, F]$ and let's note $\Delta = \frac{1}{2F}$ (the so called Nyquist frequency). Then one can recover $f$ only from discrete observations at values $f(n\Delta)$ as the Shannon-Nyquist theorem states that if $\sum_{n \in \Z} |f(n\Delta)| < +\infty$, then
$$ f(t) = \sum_{n \in Z} f(n\Delta) \frac{\sin(\pi(t/\Delta - n))}{\pi(t/\Delta - n)} $$


### Intuition for Proof 1

Since $\hat f$ has support in $I_F = [-F, F]$, assuming it is equal to its Fourier series reprensetation, one has for $x \in I_F$
\begin{split}
\hat f(x) &= \sum_{n \in \Z} c_n\left(\hat f\right) e^{i2\pi \frac{n}{2F}x} \\
&= \sum_{n \in \Z} \left( \frac{1}{2F} \int_{I_F} \hat f(t) e^{-i 2\pi \frac{n}{2F}t} dt \right)  e^{i2\pi \frac{n}{2F}x} = \sum_{n \in \Z} \left( \frac{1}{2F} \int_{\R} \hat f(t) e^{-i 2\pi \frac{n}{2F}t} dt \right)  e^{i2\pi \frac{n}{2F}x}\\
&= \sum_{n \in \Z}  \frac{1}{2F} f \left( -\frac{n}{2F} \right)   e^{i2\pi \frac{n}{2F} x} = \sum_{n \in \Z}  \frac{1}{2F} f \left( \frac{n}{2F} \right)   e^{-i2\pi \frac{n}{2F} x} \\
\end{split}

Since $\hat f$ has compact support, Fourier inversion formula yields
\begin{split}
f(t) &= \int_\R \hat f(x) e^{i 2\pi x t} dx = \int_{I_F} \hat f(x) e^{i 2\pi x t} dx \\
&= \int_{I_F}  \sum_{n \in \Z}  \frac{1}{2F} f \left( \frac{n}{2F} \right)   e^{-i2\pi \frac{n}{2F} x} e^{i 2\pi x t} dx \\
& = \sum_{n \in \Z}\frac{1}{2F} f \left( \frac{n}{2F} \right) \int_{-F}^F  e^{i 2\pi x (t-n/2F)} dx \\
& = \sum_{n \in \Z}\frac{1}{2F} f \left( \frac{n}{2F} \right) \frac{\sin\left(\pi (2Ft-n) \right)}{\pi (t-n/2F)} 
\end{split}


### Intuition for Proof 2

Let's consider a function of the form $h(t) = \int_{\R} w(x) e^{-i 2\pi xt} dx$. We do not know a priori if $h \in L^1(\R)$ and in particular, it is not assumed that $w$ is (up to sign flip) the Fourier transform of $h$.

Then, assuming sum and integral signs interchange, one has:
\begin{split}
 \frac{1}{2F} \sum_{n \in \Z} f\left(\frac{n}{2F} \right) h(t-n/2F) &= \int_{\R} \left(\frac{1}{2F} \sum_{n \in \Z} f\left(\frac{n}{2F} \right) \right) w(x) e^{-i 2\pi xt} dx \\
&= \int_{\R} \left(\frac{1}{2F} \sum_{n \in \Z} f\left(\frac{n}{2F} \right) e^{-i 2\pi \frac{n}{2F}x} \right) w(x) e^{-i 2\pi x(t-n/2F)} dx \\
&= \int_{\R} \left (\sum_{n \in \Z} \hat f(x + n2F) \right) w(x) e^{-i 2\pi x(t-n/2F)} dx 
\end{split}

Since by Poisson summation formula with $T=2F$, one has
$$ \frac{1}{2F} \sum_{n \in \Z} f\left(\frac{n}{2F} \right) e^{-i 2\pi \frac{n}{2F}x} = \sum_{n \in \Z} \hat f(x + n2F)$$

Now, note that for the very specific choice of $w(x) = 1_{I_F}(x)$, one has
$$ \left (\sum_{n \in \Z} \hat f(x + n2F) \right) w(x) e^{-i 2\pi x(t-n/2F)} = \hat f (x) 1_{I_F}(x) e^{-i 2\pi x t} = \hat f(x)e^{-i 2\pi x t} $$

since if $x \notin I_F$ then $w(x) = 0$ and if $x \in I_F$ then $\sum_{n \in \Z} \hat f(x + n2F) = \hat f(x)$ since for $n \neq 0$ we have $x +2F \notin I_F$.

In addition, for that choice of $w$, one has $h(z) = \frac{\sin(2F\pi z)}{\pi z}$, which leads to
$$ \frac{1}{2F} \sum_{n \in \Z} f\left(\frac{n}{2F} \right) \frac{\sin(\pi (t2F-n))}{\pi (t-n/2F)}  = \int_{\R} \hat f(x) e^{-i 2\pi x t} = f(t) $$


<div class="class-alert alert-block alert-info">
<b>Remark:</b> What happens if instead we choose $w(x) = 1_{I_F}(x) (1-|x|) F$ ? <br>
Notice that in this case $1_{I_F} * 1_{I_F}(x) = w(x)$ so that $\hat w(x) = F\left(\frac{\sin(2F\pi z)}{\pi z} \right)^2$
</div>

### Remarks:
- If the function $f$ is null outside of an compact interval and since (up to sign flip) $\F \hat f \sim f$, then, the sampling theorem can be applied on $\hat f$. 
- The function $sinc(z) = \frac{\sin(\pi z)}{\pi z}$ appears as a filter in the sampling theorem formula. If an observed signal is interpreted as the observation of a function sampled at a frequency $\Delta = 1/(2F)$, then the signal can be filter with $sinc(t/\Delta -n)$. The sampling formula becomes
$$ f(t) = \sum_{n\in \Z} f(n\Delta) sinc(t/\Delta - n) $$

### Impact of Truncation Error

What happens if the assumption of $\hat f$ has compact support is not satisfied and we still have the observation $f(n\Delta)$? Then we build a function $g_F(t) = \sum_{n \in Z} f(n\Delta) \frac{\sin(\pi(t/\Delta - n))}{\pi(t/\Delta - n)} $ but $f(t) \neq g(t)$. Still, the link between the 2 functions is that (under some integrability conditions)
$$ \hat g_F(t) = \sum_{n \in \Z} \hat f(t + n/\Delta)1_{[-F, F]}(t) = \hat f (t) 1_{[-F, F]}(t) $$

and therefore
$$ f(t) - g_F(t) = \int_\R \hat f(x) 1_{\{ x \notin [-F, F]\}} e^{i 2 \pi x t} dx $$

which leads to
$$ \sup_{t} |f(t) - g_F(t)| \leq \int_{\R - [-F, F]} | \hat f(x) | dx \to 0 \;\;as\; F \to +\infty$$


## An Important "Exception"

The function $t \to e^{i 2\pi xt}$ is not in $L^1(\R)$ or $L^2(\R)$ and therefore the sampling theorem does not applies. However, for all $t \in \R$, and $x \in (-M, M)$, then
$$ e^{i 2\pi xt} = \sum_{n \in \Z} e^{i 2\pi x \frac{n}{2M}} \frac{\sin\left(2 \pi M\left(t-\frac{n}{2M}\right)\right)}{2 \pi M\left(t-\frac{n}{2M}\right)}$$

# Short Time Fourier Transform

The Short Time Fourier Transform (STFT) also called Windowed Fourier Transform (WFT) is defined as:

$$ \hat f(s,x) = \int_{-\infty}^{+\infty} f(t)w(t-s) e^{-i 2 \pi tx} dt  $$

where $w$ is a window function which usually requires $w(t) \geq 0$ and $\int_{\mathbb{R}} w(t) dt < +\infty$.

For Gaussian window function used in the Gabor transform is
$$ w(x) = \frac{1}{\s \sqrt{2\pi}} e^{-\half \frac{x^2}{\s^2}} $$

The analyzing function for the STFT has two parmeter: scale and frequency as
$$ \psi_{s,x}(t) = w(t-s) e^{-i tx} $$


## Inverse STFT

$$ f(t) = c \int \int_{\mathbb{R}^2} \hat f(s,x) w(t-s) e^{i 2 \pi tx} dx ds $$



# Introduction to Wavelet Analysis

A (mother) wavelet is a function $\psi$ that satisfies:
* $\int_{-\infty}^{+\infty} \psi(x) dx = 0$
* $\int_{-\infty}^{+\infty} \frac{|\hat \psi(x)|^2}{x} dx = C_{\psi} < +\infty$

From a mother wavelet, one defines a wavelet family by scaling and translating the mother wavelet
$$ \psi_{s,x}(t) = \frac{1}{\sqrt{s}} \psi\left( \frac{t-x}{s} \right) $$

The (Continuous) Wavelet Transform is then defined by
$$ Wf(s,x) = \int_{-\infty}^{+\infty} f(t) \psi_{s,x}(t) dt $$

and can be inverted by
$$ f(t) = \frac{1}{C_{\psi}} \int_{-\infty}^{+\infty} \int_{-\infty}^{+\infty} \frac{1}{|s|^{3/2}} Wf(s,x) \psi_{s,x}(t) ds dx $$



A very simple example of such small wave is the mexican hat function:
$$ \psi(x) = \frac{2}{\pi^{1/4}\sqrt{3\s}}\left(1-\frac{x^2}{\s^2}\right) e^{-\half \frac{x^2}{\s^2}} $$


```python
def mexican_hat(x, mu, sigma):
    return 2 / (np.sqrt(3 * sigma) * np.pi**0.25) * (1 - x**2 / sigma**2) * np.exp(-x**2 / (2 * sigma**2) )

x_vals = np.arange(-10,10,0.1)
plt.plot(x_vals, mexican_hat(x_vals, 0, 1))
plt.show()
```
